<h1 style="color: blue;">가. CNN 소개</h1>

<h2 style="color: green;">CNN의 필요성과 특징</h2>

복잡한 구조와 단순한 머신러닝의 한계를 극복하기 위해 DNN이 많이 쓰였다. 하지만 DNN은 입력층에 1차원 데이터를 넣어주어야하기 때문에 flatten을 시켜주어 학습을 진행하였다. 이미지와 동영상은 다차원 형태로 이루어져있는 데이터 형태이기 때문에 flatten과정에서 공간적/지역적 정보가 손실되게 된다. 또한 추상화과정이 없어 학습시간과 능률의 효율성이 저하된다.

CNN은 이미지를 날것(raw input) 그대로 받음으로써 공간적/지역적 정보를 유지한 채 특성(feature)들의 계층을 빌드업합니다. CNN의 중요 포인트는 이미지 전체보다는 부분을 보는 것, 그리고 이미지의 한 픽셀과 주변 픽셀들의 연관성을 살리는 것입니다.

<h2 style="color: green;">CNN의 구조</h2>

### **입력층 -> 합성공층 -> 활성화함수(은닉층) -> 풀링층 -> 활성화함수(은닉층)** ->  완전 연결층 -> 출력층

+ 합성곱층 : 합성곱층은 입력 이미지에 대해 합성곱 연산을 수행하여 특징 맵을 생성합니다. 합성곱 연산은 커널(필터)과 입력 이미지의 일부 영역을 곱하고 모두 더하는 연산을 의미합니다. 이를 통해 입력 이미지의 공간적인 특징을 추출합니다. 합성곱층은 여러 개의 필터를 사용하여 여러 개의 특징 맵을 생성할 수 있습니다.

+ 활성화 함수(Activation Function): 합성곱층 이후에는 활성화 함수를 적용합니다. 일반적으로는 ReLU(Rectified Linear Unit) 함수가 사용됩니다. 활성화 함수는 특징 맵의 선형 조합에 비선형성을 추가하여 모델의 표현력을 향상시킵니다.

+ 풀링층(Pooling Layer): 풀링층은 특징 맵의 크기를 줄이고 추상화된 정보를 유지하는 역할을 합니다. 일반적으로 최대 풀링(Max Pooling)이 사용되는데, 이는 특정 영역에서 최댓값을 추출하여 특징을 보존하면서 공간적인 크기를 줄이는 것을 의미합니다.

+ 완전 연결층(Fully Connected Layer): 완전 연결층은 풀링층의 출력을 받아서 최종적인 분류나 예측을 수행합니다. 이 층은 모든 뉴런이 이전 층의 모든 뉴런과 연결되어 있는 구조를 가지고 있습니다. 활성화 함수를 적용하여 최종 출력을 얻게 됩니다.

<h2 style="color: green;">1. 합성곱 계층</h2>




**커널(필터)**라고 부르는 것은 전체 이미지를 임의의 값으로 필터링하는 도구이다. 전체 이미지를 공간별로 읽어오면 합성곱을 통해 특징을 수치로 반환한다. 이렇게 필터값을 설정하여 하나의 전체 이미지를 데이터를 분석하여 여러 맵을 생성해 낼 수 있다.

*합성곱은 행렬의 같은 위치에 있는 요소를 곱하여 모두 더해주어서 계산한다. 행렬곱셈과는 연산방법이 다르다*

**합성곱 레이어(Convolutional Layer)는 Convolutional Neural Network (CNN)에서 가장 중요한 역할을 수행하는 레이어입니다. 합성곱 레이어는 입력 데이터로부터 특징을 추출하는 과정을 담당하며, 이미지 처리와 패턴 인식에 탁월한 성능을 보입니다.**

+ 특징 추출: 합성곱 레이어는 입력 데이터로부터 특징을 추출합니다. 이를 위해 커널 또는 필터(filter)라고 하는 작은 크기의 가중치 행렬을 사용하여 입력 데이터를 스캔합니다. 이 과정에서 커널과 입력 데이터의 합성곱 연산이 수행되며, 특정한 패턴이나 특징을 감지하는 역할을 합니다.
<br></br>
+ 지역적인 특징 인식: 합성곱 레이어는 지역적인 특징을 인식할 수 있습니다. 이는 커널이 입력 데이터를 작은 영역 단위로 스캔하면서 해당 영역의 패턴을 파악하는 것을 의미합니다. 이러한 지역 특징 인식은 이미지의 공간적인 구조를 파악하고, 이미지 내에서의 패턴 인식을 용이하게 합니다.
<br></br>
+ 특징 맵 생성: 합성곱 레이어는 여러 개의 커널을 동시에 사용하여 다양한 특징을 추출합니다. 이 과정에서 입력 데이터의 각 영역별로 커널과의 합성곱이 수행되고, 결과로 여러 개의 특징 맵(feature map)이 생성됩니다. 각 특징 맵은 입력 데이터의 다른 측면이나 특성을 나타내며, 다음 레이어로 전달됩니다.
<br></br>
+ 파라미터 공유: 합성곱 레이어는 커널을 이용한 합성곱 연산을 통해 특징을 추출하므로, 커널의 가중치를 학습하는 과정이 필요합니다. 하지만 합성곱 레이어에서는 커널의 파라미터를 입력 데이터의 모든 위치에서 동일하게 공유하여 사용합니다. 이를 통해 학습해야 할 파라미터의 수를 대폭 줄이고, 모델의 효율성과 일반화 성능을 향상시킵니다.

<h2 style="color: green;">Padding</h2>

### *패딩의 목적*

합성곱 신경망에서 입력 데이터의 공간적인 크기를 보존하거나 조정하는 기술입니다. 합성곱 연산을 수행할 때 입력 데이터 주위에 추가적인 값으로 채워진 빈 영역을 만들어줍니다. 이렇게 패딩을 적용함으로써 입력 데이터의 크기를 유지하거나 적절히 조절할 수 있습니다.

+ 경계 처리: 합성곱 연산은 입력 데이터의 가장자리에 위치한 픽셀 또는 요소들의 정보를 손실할 수 있습니다. 패딩을 적용함으로써 입력 데이터의 경계 부분에 대한 정보를 보존하고, 모델이 가장자리에 위치한 패턴을 학습할 수 있도록 도와줍니다.
<br></br>
+ 출력 크기 조절: 합성곱 연산은 입력과 출력의 공간적인 크기를 조절하는 역할을 합니다. 패딩을 사용하여 입력 데이터의 크기를 조절함으로써, 출력 데이터의 크기를 원하는 대로 조절할 수 있습니다.
<br></br>
+ 공간 정보 유지: 패딩을 사용하여 입력 데이터의 공간 정보를 유지할 수 있습니다. 합성곱 신경망은 주로 이미지와 같은 2D 또는 3D 공간 데이터를 처리하는 데 사용되며, 패딩을 통해 입력 데이터의 공간 구조를 유지하면서 특징을 추출할 수 있습니다.

### *Valid Padding*

패딩없이 순수한 입력배열에서만 합성곱을 적용하여 맵을 만드는 것을 Valid Padding이라고 한다. 
입력 데이터 주위에 추가적인 값으로 채워지는 패딩을 적용하지 않기 때문에 출력 크기는 입력 크기보다 작아진다. 때문에 입력 데이터의 공간적인 구조를 보존하기 어렵게 만든다.



또한 이처럼 경계효과가 발생되어 끝에 있는 요소값들은 중심부에 비해서 합성곱 연산에 비교적 적게 참여하게 되면서 가장자리 특징을 잘 못잡게 된다. Valid Padding에서는 이러한 경계효과를 방지하기 위해서 가장자리에 있는 원소들을 합성곱 연산에 참여시키지 않는다. 이로 인해ㅔ 입력 데이터의 가장자리에 있는 중요한 특징이나 패턴이 무시되지 않고 보존된다. 하지만 경계부분에 대해서 학습을 하지 못해 정보손실이 발생 할 수 있다. 때문에 우리는 **Padding**이라는 과정을 거쳐서 이미지의 효율적인 학습을 할 수 있다.

### *Zero Paddiing*



0로 구성된 테두리를 이미지 가장자리에 감싸 준다고 생각하면 됩니다. 5x5크기의 이미지 입력값이 7x7크기가 될겁니다. 이 상태에서 3x3필터를 적용해줄 경우 똑같이 5x5크기의 결과값을 얻을수 있게 된다. 이렇게 되면 입력값의 크기와 결과값의 크기가 같아졌음, 즉 손실이 없어졌음을 볼 수 있습니다.
이 과정에서 입력 데이터의 크기를 일정하게 유지시키기 위해서 패딩을 적용하는 것을 **Same Padding**이라고 한다. 

네트워크 구성의 단순화: 세임 페딩을 적용하면 출력 크기가 입력 크기와 동일하게 유지되므로, 네트워크의 구성을 단순화할 수 있습니다.

효율적인 학습: 세임 페딩을 사용하면 입력과 출력의 크기가 동일하므로, 출력과의 오차를 계산하여 역전파를 통해 효과적으로 학습할 수 있습니다.

**Zero Padding의 효과**

+ 보존된 공간 정보: Zero Padding은 입력 데이터의 가장자리에 0 값을 추가하여 입력과 출력의 공간적인 구조를 보존합니다. 이는 모델이 입력 데이터의 가장자리에 있는 중요한 패턴이나 특징을 인식할 수 있도록 합니다.

+ 경계 처리 개선: Zero Padding은 경계 부분에서도 합성곱 필터의 중심과의 겹치는 영역을 늘립니다. 이는 경계 부분에서도 특징을 정확하게 인식하고 활용할 수 있도록 도와줍니다.

+ 출력 크기 보존: Zero Padding은 입력 데이터의 가장자리에 0 값을 추가하는 것이므로, 출력 크기를 유지할 수 있습니다. 이는 모델의 다음 레이어로 전달되는 정보의 양을 보존하면서 더 많은 특징을 학습할 수 있도록 합니다.

+ 모델의 안정성 개선: Zero Padding은 입력 데이터의 가장자리에 0 값을 추가하여 모델의 안정성을 개선할 수 있습니다. 이는 입력 데이터의 작은 변화에 민감하게 반응하는 것을 완화하고, 모델이 입력에 더 일반화된 특징을 학습할 수 있도록 도움을 줍니다.

**Zero Padding의 단점**

+ 메모리 사용량 증가: Zero Padding은 입력 데이터의 가장자리에 0 값을 추가하기 때문에 메모리 사용량이 증가합니다. 특히 큰 크기의 입력 데이터와 깊은 신경망에서는 많은 수의 0 값이 추가되므로 메모리 요구사항이 더 커질 수 있습니다.

+ 연산량 증가: Zero Padding은 추가된 0 값을 계산해야 하기 때문에 연산량이 증가합니다. 특히 큰 크기의 입력 데이터에서는 이러한 연산량 증가가 모델의 전체적인 속도에 영향을 줄 수 있습니다.

+ 공간적인 정보 왜곡: Zero Padding은 입력 데이터의 가장자리에 0 값을 추가하는 것이기 때문에 입력 데이터의 공간적인 정보가 왜곡될 수 있습니다. 이는 모델이 입력 데이터의 가장자리에 있는 패턴이나 특징을 불필요하게 강조할 수도 있습니다.

+ 경계 처리 문제: Zero Padding은 입력 데이터의 가장자리에 0 값을 추가하는 방식이므로 경계 부분의 처리에 일부 제약이 있을 수 있습니다. 특히 입력 데이터의 가장자리에 있는 특징이 모델에 제대로 전달되지 않을 수 있습니다.

<h2 style="color: green;">Stride</h2>

Stride는 쉽게 말해 필터를 얼마만큼 움직여 주는가에 대한 것입니다. stride값이 1일경우 필터를 한칸씩 움직여 준다고 생각하면 됩니다. 참고로 stride-1이 기본값이고 stride는 1보다 큰 값이 될 수도 있습니다. stride값이 커질 경우 필터가 이미지를 건너뛰는 칸이 커짐을 의미하므로 결과값 이미지의 크기는 작아짐을 의미합니다.

**Stride 장점**

+ 계산량 감소: 스트라이드를 증가시킴으로써 필터가 입력 데이터를 건너뛰어 스캔하는 비율을 늘릴 수 있습니다. 이는 합성곱 연산에 참여하는 연산량을 줄여 계산 비용을 감소시킬 수 있습니다. 따라서 모델의 효율성과 속도를 향상시킬 수 있습니다.

+ 공간 축소: 스트라이드를 증가시키면 출력 특징 맵의 크기가 입력 데이터의 크기에 비해 작아집니다. 이는 공간적인 정보를 다소 손실하게 되지만, 특징 추출을 위한 필터의 적용 범위가 넓어짐을 의미합니다. 이는 더 넓은 영역의 특징을 인식할 수 있게 해주고, 추상화된 특징을 잘 포착할 수 있습니다.

**Stride 단점**

+ 공간 해상도 감소: 스트라이드를 증가시키면 출력 특징 맵의 공간 해상도가 감소합니다. 작은 스트라이드를 사용하는 경우에는 입력 데이터의 세부적인 패턴과 특징을 잘 보존할 수 있지만, 큰 스트라이드를 사용하면 입력 데이터의 공간적인 세부 정보가 희석될 수 있습니다.

+ 공간적인 이동 불가능성: 스트라이드가 입력 데이터의 크기보다 큰 경우, 필터가 입력 데이터를 완전히 커버하지 못하고 데이터의 일부만 스캔하게 됩니다. 이는 입력 데이터의 모든 공간적인 정보를 활용하지 못하고, 특징을 제대로 인식하지 못할 수 있습니다.

스트라이드는 합성곱 신경망의 설계와 성능 조정에 있어서 중요한 하이퍼파라미터입니다. 최적의 스트라이드 값을 선택하는 것은 모델의 성능과 특징 추출 능력에 영향을 미치므로, 실험과 조정을 통해 적절한 값을 찾아내어야 합니다.
