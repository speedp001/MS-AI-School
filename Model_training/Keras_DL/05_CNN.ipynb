```python
#CNN 합성공 신경망

from keras import models
from keras import layers

model = models.Sequential()
#첫 번째 레이어에만 input_shape을 하면 된다. 이후는 더 작지만 않으면 오류 안나고 컴파일된다.
model.add(layers.Conv2D(32,(3,3), activation='relu', input_shape=(28,28,1)))
#필터의 크기(3,3), 필터의 개수(각 필터는 값이 행렬값이 다르다)
model.add(layers.MaxPooling2D((2,2)))
#MaxPooling(2,2)는 풀링할 구역 크기(각 구역에서 가장 큰 값만 뽑아내서 새로운 행렬 생성)
model.add(layers.Conv2D(32,(3,3), activation='relu', input_shape=(28,28,1)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(32,(3,3), activation='relu', input_shape=(28,28,1)))
model.add(layers.MaxPooling2D((2,2)))
```

```python
model.summary()
```

```python
#2차원 데이터를 평탄화하는 작업
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
#이전 단계에서 데이터값을 확률분포로 변환해주는 층
model.add(layers.Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
from keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

#데이터 전처리 과정
#CNN 신경망에 넣기위해서 input_shape과 동일하게 차원을 맞춘다.
train_images.shape
train_data = train_images.reshape((60000,28,28,1))  #1이면 흑백 3이면 컬러
#픽셀값을 255로 나누어서 0과 1 사이로 통일
train_data = train_data.astype("float32") / 255


test_data = test_images.reshape((10000,28,28,1))
test_data = test_data.astype("float32") / 255

#Label 데이터의 이진 벡터화
train_target = to_categorical(train_labels)
test_target = to_categorical(test_labels)
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_data, train_target, epochs=5, batch_size=64)
```

```python
test_loss, test_acc = model.evaluate(test_data, test_target)
```
