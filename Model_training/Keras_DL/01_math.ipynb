```python
#논리 게이트

#AND
def AND(a,b):
  input = np.array([a,b])
  weights = np.array([0.4,0.4])
  bias = -0.6
  value = np.sum(input * weights) + bias
  #sum은 모든 배열의 값을 더한다.

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
#OR
def OR(a,b):
  input = np.array([a,b])
  weights = np.array([0.4,0.4])
  bias = -0.3
  value = np.sum(input * weights) + bias
  #sum은 모든 배열의 값을 더한다.

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
#NAND
def NAND(a,b):
  input = np.array([a,b])
  weights = np.array([-0.5,-0.5])
  bias = 0.7
  value = np.sum(input * weights) + bias
  #sum은 모든 배열의 값을 더한다.

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
#Step Function (계단함수)

def step_function(x):
  if x>0:
    return 1
  else:
    return 0
```

```python
print(step_function(0))
print(step_function(-1))
print(step_function(0.1))
```

```python
def step_function_for_numpy(x):
  y= x>0
  return y.astype(np.int)
  #True는 1 False는 0으로 반환
```

```python
a = np.array([5.3,-4.2,0])
print(step_function_for_numpy(a))
```

```python
#Sigmoid Function

def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(10))
print(sigmoid(100))
print(sigmoid(-3))
print(sigmoid(-100))
```

```python
#Step Function vs Sigmoid Function

x = np.arange(-5,5,0.01)
y1 = sigmoid(x)
y2 = step_function_for_numpy(x)

plt.plot(x, y1, 'r-',x ,y2, 'b-')
```

```python
#Softmax Function

def softmax(x):
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  y = exp_x / sum_exp_x
  
  return y
```

```python
x = np.array([0.3,0.2,3.0,-1,2])
print(softmax(x))
```

```python
print(np.sum(softmax(x)))
```
