<h2 style="color: green;">앙상블 기법 - 부스팅</h2>



부스팅(Boosting)은 앙상블 학습(Ensemble Learning)의 일종으로, 약한 학습기(Weak Learner)를 순차적으로 학습시켜 강한 학습기(Strong Learner)를 만드는 방식입니다. 부스팅은 이전 학습기의 오차를 보완하는 방식으로 동작하여, 이전 학습기에서 잘못 분류된 샘플에 가중치를 부여하고 다음 학습기에서 이를 보정 합니다. 이를 반복하면서 모든 학습기를 결합하여 최종 예측을 수행합니다.

### *부스팅 동작방식*

1. 초기화 : 모든 학습 데이터에 동일한 가중치를 할당합니다.
<br></br>
2. 학습기 학습 : 첫 번째 학습기를 초기 데이터와 가중치를 사용하여 학습시킵니다.
<br></br>
3. 예측 및 가중치 갱신 : 첫 번째 학습기를 사용하여 예측을 수행하고, 예측 결과와 실제값의 차이를 기반으로 샘플의 가중치를 조정합니다. 잘못 분류된 샘플은 가중치를 증가시키고, 올바르게 분류된 샘플은 가중치를 감소시킵니다.
<br></br>
4. 다음 학습기 학습 : 가중치가 조정된 데이터를 사용하여 다음 학습기를 학습시킵니다. 이전 학습기의 오차에 영향을 받은 샘플들에 더욱 집중하여 학습이 이루어집니다.
<br></br>
5. 반복 : 3번부터 4번까지의 과정을 지정된 학습기 개수나 정지 조건을 만족할 때까지 반복합니다.
<br></br>
6. 예측 결합 : 모든 학습기의 예측을 결합하여 최종 예측을 수행합니다. 예측 결과는 보통 가중 평균이나 다수결 등의 방식으로 조합됩니다.

**이렇게 부스팅 알고리즘은 순차적인 학습과 가중치 갱신을 통해 이전 학습기의 오차를 보정하며, 강한 학습기를 만들어냅니다. 각 학습기는 이전 학습기에서 잘못 분류된 데이터에 집중하여 학습하므로, 전체 앙상블의 성능이 향상됩니다.**


### *AdaBoost 알고리즘*

에이다부스트(AdaBoost)는 앙상블 학습(Ensemble Learning) 알고리즘 중 하나로, 약한 학습기(Weak Learner)들을 결합하여 강력한 분류기를 구축하는 방법입니다. 에이다부스트는 예측력이 약한 학습기들을 여러번 반복적으로 학습시켜 각각의 학습기가 잘못 분류한 샘플들에 가중치를 부여하여 다음 학습기에 집중적으로 학습시킵니다.

에이다부스트는 일반적으로 다른 앙상블 학습 알고리즘들과 비교하여 학습 속도가 빠르고 예측 성능이 높은 편입니다. 하지만 데이터에 이상치(Outlier)가 많거나 노이즈가 포함된 경우, 오분류된 샘플들에 대한 가중치가 너무 높아져 과적합(Overfitting)되는 문제가 발생할 수 있습니다.
