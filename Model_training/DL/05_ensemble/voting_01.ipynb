<h2 style="color: green;">앙상블 기법 - 보팅</h2>



**보팅(Voting)은 앙상블 학습(Ensemble Learning)의 한 종류로, 여러 개의 다른 예측 모델들로부터 예측 결과를 결합하여 최종 예측을 수행하는 방법입니다. 보팅은 다양한 모델의 다양성을 활용하여 더 강력하고 안정적인 예측을 할 수 있는 장점이 있습니다.**

*주의사항 : 보팅은 여러 모델의 예측을 결합하는 것이고, 배깅은 동일한 모델을 다른 데이터 샘플에 대해 학습하여 예측을 결합하는 것입니다.*

### *다양한 보팅 기법 - 하드 보팅(Hard Voting)*

하드 보팅 (Hard Voting) : 하드 보팅은 다수의 모델이 예측한 결과를 다수결 방식으로 결합하여 최종 예측을 수행하는 방법입니다. 각 개별 모델은 클래스 레이블 중에서 가장 빈도가 높은 클래스를 선택하고, 그 선택들 중에서 다수결로 최종 예측 클래스를 결정합니다. 이는 이진 분류에서는 0 또는 1과 같이 확실한 클래스로 예측하는 데 사용될 수 있습니다. 예를 들어, 세 개의 모델이 0, 1, 1을 예측했다면 하드 보팅에 따라 다수결로 최종 예측은 1이 됩니다.

1. 하드 보팅은 다수결 원칙을 사용하여 최종 예측을 결정합니다.
2. 각각의 개별 모델들이 클래스 레이블에 대한 예측을 하고, 이들 예측 중 가장 많은 표를 받은 클래스를 최종 예측으로 선택합니다.

### 다양한 보팅 기법 - 소프트 보팅(Soft Voting)

소프트 보팅 (Soft Voting) : 소프트 보팅은 다수의 모델이 예측한 클래스 확률을 평균하여, 가장 높은 평균 확률을 가진 클래스를 최종 예측으로 선택하는 방법입니다. 각 개별 모델은 클래스 확률을 예측하고, 이들 확률을 평균하여 가장 높은 평균 확률을 가지는 클래스를 최종 예측 클래스로 결정합니다. 이는 클래스 확률을 제공하는 모델에 유용하며, 각 모델이 클래스에 대한 상대적 신뢰도를 반영하는 데 사용될 수 있습니다. 예를 들어, 세 개의 모델이 클래스 0의 확률을 [0.3, 0.4, 0.5]로, 클래스 1의 확률을 [0.7, 0.6, 0.5]로 예측했다면, 소프트 보팅에 따라 평균 확률이 높은 클래스 1이 최종 예측 클래스가 됩니다.

1. 소프트 보팅은 개별 모델들의 예측값들의 평균 또는 확률을 사용하여 최종 예측을 결정합니다.
2. 분류(Classification)문제의 경우 각 클래스에 대한 예측 확률을 평균하여 가장 높은 확률을 가진 클래스를 선택합니다.
3. 회귀(Regression)문제의 경우 개별 모델들의 예측값들의 평균을 사용합니다.

### 다양한 보팅 기법 - 가중치 보팅(Weighted Voting)

1. 가중치 보팅은 각 개별 모델에 가중치를 부여하여 예측 결과를 결합합니다.
2. 일부 모델들이 더 신뢰할 수 있다고 판단되는 경우, 해당 모델에 더 높은 가중치를 할당하여 최종 예측에 더 큰영향을 줄 수 있습니다.

### 다양한 보팅 기법 - 배깅(Bagging)

1. 배깅은 복원추출(ResamplingwithReplacement)을 사용하여 다양한 훈련 데이터셋을 생성하고, 각 데이터셋에 대해 개별 모델을 학습시키는 방식입니다.
2. 개별 모델들의 예측 결과를 보팅을 통해 결합하여 최종 예측을 수행합니다.
3. 대표적인 배깅 알고리즘으로는 랜덤포레스트(RandomForest)가 있습니다.

### 다양한 보팅 기법 - 부스팅(Boosting)

1. 부스팅은 약한 학습기(WeakLearner)들을 순차적으로 학습시켜 강한학습기(StrongLearner)를만드는 방식입니다.
2. 각 모델들은 이전 모델의 오차를 보완하기위해 가중치를 조절하면서 학습됩니다.
3. 대표적인 부스팅 알고리즘으로는 에이다부스트(AdaBoost), 그래디언트부스(GradientBoosting)등이 있습니다.

### 보팅 모델 구성과 성과 분석

1. 기본 모델 구성 : 앙상블 모델을 구성하기 위해 다양한 기본 모델들을 선택합니다. 기본 모델은 독립적으로 학습되는 모델들로, 서로 다른 알고리즘, 구조, 하이퍼파라미터 등을 가질 수 있습니다. 일반적으로 기본 모델들은 서로 다른 특성을 갖고있어야 앙상블의 효과를 극대화할 수 있습니다.
<br></br>
2. 데이터 분할 : 데이터를 학습 세트와 테스트 세트로 분할합니다. 학습 세트를 사용하여 기본 모델들을 개별적으로 학습시키고, 테스트 세트를 사용하여 모델의 성능을 평가합니다.
<br></br>
3. 기본 모델 학습 : 각 기본 모델은 개별적으로 학습됩니다. 학습 세트를 사용하여 모델을 학습시키고, 검증 세트 등을 사용하여 모델의 성능을 조정할 수 있습니다. 이를 통해 각 기본 모델은 자체적인 예측 능력을 갖게 됩니다.
<br></br>
4. 보팅 앙상블 모델 구성: 기본 모델들의 학습이 완료되면, 보팅 앙상블 모델을 구성합니다. 보팅은 하드 보팅 또는 소프트 보팅으로 수행될 수 있습니다. 하드 보팅은 다수결로 최종 예측을 결정하고, 소프트 보팅은 클래스 확률을 평균하여 최종 예측을 결정합니다.
<br></br>
5. 앙상블 모델 성과 분석: 보팅 앙상블 모델을 사용하여 테스트 세트에 대한 예측을 수행합니다. 이를 기반으로 정확도, 정밀도, 재현율, F1 스코어 등과 같은 성과 지표를 계산하여 모델의 성능을 평가합니다. 또한 개별 기본 모델들의 성능과 앙상블 모델의 성능을 비교하여 앙상블의 효과를 확인할 수 있습니다.

**앙상블 모델의 성과 분석은 기본 모델들의 예측이 상호보완적이고 다양성을 가지는 경우에 가장 효과적입니다. 여러 기본 모델들이 독립적으로 예측을 수행하므로, 오차가 서로 상호 보완되어 예측의 안정성이 향상됩니다. 앙상블 모델은 단일 모델보다 높은 정확도와 일반화 성능을 보일 수 있습니다. 또한, 앙상블 모델은 개별 모델보다 더욱 견고하고 안정적인 예측을 제공할 수 있어 신뢰성이 높습니다.**
