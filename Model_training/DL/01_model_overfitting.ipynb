<h2 style="color: green;">부적합과 과적합</h2>

### 부적합
train, test 데이터셋 모두에서 성능이 떨어지는 경우

**특징**

+ 모델이 학습 데이터에서도 낮은 정확도를 보입니다.
<br></br>
+ 모델이 너무 단순하거나 제한된 특성을 학습하므로, 실제 데이터의 복잡한 패턴을 잡아내지 못합니다.
<br></br>
+ 과소적합된 모델은 학습 데이터와 테스트 데이터에서 모두 성능이 낮아집니다.

### 과적합
train, valid 데이터셋 성능이 우수하나 test 에서 떨어지는 경우

**특징**

+ 학습 데이터에서는 높은 정확도를 보이지만, 검증 데이터나 테스트 데이터에서는 성능이 낮아집니다. 
<br></br>
+ 모델의 복잡도가 높아질수록 과적합의 위험이 증가합니다.
<br></br>
+ 과적합된 모델은 새로운 데이터에 대한 일반화 능력이 떨어져서 실제 환경에서의 성능이 저하됩니다.

<h2 style="color: green;">드롭아웃(Dropout)</h2>

**드롭아웃(Dropout)** 은 딥러닝 모델에서 과적합을 방지하기 위한 정규화(regularization) 기법 중 하나입니다. 드롭아웃은 학습 과정에서 신경망의 일부 유닛을 랜덤하게 제거하여 모델의 일반화 능력을 향상시키는 방법입니다.



드롭아웃은 학습 단계에서 신경망의 일부 유닛을 임의로 제거하여, 제거된 유닛을 포함한 신경망의 부분집합으로 학습을 수행합니다.

각 학습 단계에서 제거되는 유닛은 확률적으로 결정되며, 보통 0.2 에서 0.5 사이의 확률로 선택됩니다. 이렇게 제거된 유닛은 그래디언트를 역전파할 때 무시되고, 가중치 업데이트에도 영향을 주지 않습니다.

### 동작 방식

1. 학습 단계에서, 드롭아웃은 각 유닛의 출력을 보존할 확률(keep probability)에 따라 임의로 선택된 유닛을 제거합니다. 제거된 유닛은 해당 단계에서 신호 전달과 가중치 업데이트에 참여하지 않습니다.
<br></br>
2. 제거되지 않은 유닛은 보존된 확률에 따라 스케일링되어 출력됩니다. 이는 제거된 유닛의 활성화 값이 보존된 유닛들에 고르게 분산되도록 보장합니다.
<br></br>
3. 역전파 단계에서, 드롭아웃은 제거된 유닛에 대한 그래디언트 계산을 생략하여 모델의 파라미터 업데이트에 영향을 주지 않습니다.

### 드롭아웃 장점

+ 과적합을 방지하고 모델의 일반화 능력을 향상시킵니다.
<br></br>
+ 앙상블 효과를 얻어 다양한 서브넷(subnetworks)으로 학습하여 모델의 안정성을 향상시킵니다.
<br></br>
+ 파라미터의 과도한 의존성을 줄여 신경망을 더 견고하게 만듭 니다.

**드롭아웃 개념과 동작 방식드롭아웃은 주로 합성곱 신경망(Convolutional Neural Networks)과 전결합층 신경망(Fully Connected Neural Networks)에서 사용되며, 과적합을 방지하고 일반화 성능을 향상시키는 데 효과적입니다.**

### 드롭아웃 한계

+ 정보 손실: 드롭아웃은 학습 과정에서 일부 유닛을 무작위로 제거하여 학습 데이터의 다양성을 증가시킵니다. 그러나 이는 모델이 해당 유닛에서 가지고 있는 정보를 완전히 잃을 수도 있음을 의미합니다. 따라서 유의미한 정보가 있는 유닛이 무작위로 제거될 경우, 모델의 성능이 저하될 수 있습니다.
<br></br>
+ 계산 비용: 드롭아웃은 학습 단계에서 유닛을 제거하고 다시 활성화하는 과정을 포함하므로 계산 비용이 추가로 발생합니다. 특히 큰 모델이나 복잡한 신경망에서는 계산 비용이 상당할 수 있습니다. 따라서 드롭 아웃을 사용하면 모델의 훈련 속도가 저하될 수 있습니다.
<br></br>
+ 하이퍼파라미터 설정의 어려움: 드롭아웃에서는 제거할 유닛의 비율(keep probability)을 정해야합니다. 이 비율은 모델의 성능에 직접적인 영향을 미치며, 최적의 비율은 문제에 따라 다를 수 있습니다. 따라서 적절한 keep probability 값을 찾기 위해서는 실험과 조정이 필요할 수 있습니다.
<br></br>
+ 시간적 일관성의 손실: 드롭아웃은 학습 단계에서 유닛을 무작위로 제거하여 모델을 정규화합니다. 그러나 예측 단계에서는 드롭아웃을 적용하지 않습니다. 따라서 학습과 예측 단계 사이에 시간적 일관성이 손실될 수 있습니다. 이는 일부 응용 프로그램(예: 시계열 데이터)에서 성능에 영향을 줄 수 있습니다.

## *드롭아웃(Dropout) 실습*

```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

from torch.utils.data import DataLoader
```

### *드롭아웃 적용한 신경망 모델 정의*

```python
class DropoutNet(nn.Module) :
    def __init__(self) :
        super(DropoutNet, self).__init__()
        self.fc1 = nn.Linear(784, 500)
        self.dropout = nn.Dropout(p=0.5)
        self.fc2 = nn.Linear(500, 10)
        
    def forward(self, x) :
        #x.size(0) = 2 -> x.view(2, -1) -> 2차원인 배열을 아무형태로 변환하여라 -> 결과값 1차원으로 평탄화
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        #fc1레이어에 dropout 적용
        x = self.fc2(x)
        
        return x

# 드롭아웃 적용하지 않은 신경망 모델 정의
class NonDropoutNet(nn.Module) :
    def __init__(self) :
        super(NonDropoutNet, self).__init__()
        self.fc1 = nn.Linear(784, 500)
        self.fc2 = nn.Linear(500, 10)
        
    def forward(self, x) :
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        
        return x
```

```python
train_transform = transforms.Compose([
    transforms.AutoAugment(),
    transforms.ToTensor(),
    # 정규화할 채널 수 만큼 요소 개수가 정해진다.
    transforms.Normalize((0.5,), (0.3,))
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.3,))
])

train_dataset = datasets.MNIST("../data/", train=True, download=True, transform=train_transform)
test_dataset = datasets.MNIST("../data/", train=False, download=True, transform=test_transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
```

### *드롭아웃 적용 모델 초기화, loss, 옵티마이저 정의*

```python
# 드롭아웃 적용 모델 초기화
dropout_model = DropoutNet()
print(dropout_model)

# loss 함수, 옵티마이저 설정
dropout_criterion = nn.CrossEntropyLoss()
dropout_optimizer = optim.SGD(dropout_model.parameters(), lr=0.0001)
```

### *드롭아웃 모델 학습 및 평가*

```python
# 드롭아웃 적용 모델 학습, 평가 정의

# 드롭아웃 적용 모델 학습 코드
for epoch in range(10) :
    dropout_model.train()  #train모드 선언
    for images, labels in train_loader :
        dropout_optimizer.zero_grad()
        dropout_output = dropout_model(images)
        dropout_loss = dropout_criterion(dropout_output, labels)
        dropout_loss.backward()
        dropout_optimizer.step()
        
# 드롭아웃 적용 모델 평가 코드
dropout_model.eval()
with torch.no_grad() :
    dropout_correct = 0
    dropout_total = 0
    
    for images, labels in test_loader :
        test_out = dropout_model(images)
        # print(test_out)
        # test_out에서 각 숫자 0~9까지 모델을 통해 계산된 확률이 분포
        # print(torch.max(test_out.data, 1))
        # 그 확률 분포 중 가장 확률이 높은 값과 그에 맞는 인덱스를 반환
        _, dropout_pre = torch.max(test_out.data, 1)
        # .data를 사용하는 이유는 기울기 추적 등에 영향을 주지 않으면서 좀 더 직접적인 데이터 접근이 필요할 때 사용하기 위해서입니다.
        # 1을 전달하면 텐서의 두 번째 차원에서 최댓값을 찾아 해당 인덱스를 반환(test_out 데이터는 2차원데이터이기 때문)
        dropout_total += images.size(0)
        dropout_correct += (dropout_pre == labels).sum().item()
        
    print("드롭아웃 적용 모델 정확도 >> {:.2f}%".format(100 * dropout_correct / dropout_total))
```

### *드롭아웃 적용되지 않은 모델 학습 평가 정의*

```python
# 모델 호출
non_dropout_model = NonDropoutNet()

# loss 함수, 옵티마이저 설정
non_dropout_criterion = nn.CrossEntropyLoss()
non_dropout_optimizer = optim.SGD(non_dropout_model.parameters(), lr=0.0001)

# 드롭아웃 적용되지 않은 모델 학습
for epoch in range(10) :
    non_dropout_model.train()  #train모드 선언
    for images, labels in train_loader :
        non_dropout_optimizer.zero_grad()
        non_dropout_output = non_dropout_model(images)
        non_dropout_loss = non_dropout_criterion(non_dropout_output, labels)
        non_dropout_loss.backward()
        non_dropout_optimizer.step()
        
# 드롭아웃 적용 모델 평가 코드
non_dropout_model.eval()
with torch.no_grad() :
    non_dropout_correct = 0
    non_dropout_total = 0
    
    for images, labels in test_loader :
        non_test_out = non_dropout_model(images)
        _, non_dropout_pre = torch.max(non_test_out.data, 1)
        non_dropout_total += images.size(0)
        non_dropout_correct += (non_dropout_pre == labels).sum().item()
        """
        .item()은 텐서가 단일 값을 가지는 경우에만 사용할 수 있습니다. 
        따라서, (non_dropout_pre == labels).sum().item()은 비교한 결과 True인 원소들의 개수를 추출한 후, 
        해당 값을 스칼라 값으로 변환하여 반환합니다.
        """
        
    print("드롭아웃 적용되지 않은 모델 정확도 >> {:.2f}%".format(100 * non_dropout_correct / non_dropout_total))
```

<h2 style="color: green;">라벨 스무딩</h2>

### 라벨 스무딩 소개

**오버피팅(과적합) 문제를 해결하기 위한 기법 중 하나로 "라벨 스무딩(Label Smoothing)"이 있습니다. 라벨 스무딩은 모델이 학습할 때 입력 데이터의 정확한 라벨 대신에 부분적으로 다른 라벨 정보를 주는 방법입니다.**

기존의 분류 문제에서는 각 입력 데이터에 대해 하나의 정확한 라벨을 가지고 있습니다. 그러나 이렇게 엄격하게 정확한 라벨을 제공하면 모델은 주어진 라벨에 완벽하게 맞추려는 경향이 생깁니다. 이는 모델이 훈련 데이터에 과도하게 적합되고, 실제로는 일반화하기 어려운 패턴을 학습하는 과적합 문제를 야기할 수 있습니다.

라벨 스무딩은 이러한 과적합 문제를 완화시키기위해 사용됩니다. 대신에 모든 라벨을 부분적으로 다른 라벨로 대체하는 것입니다. 예를 들어, 일반적으로 각 클래스의 라벨은 0 또는 1로 표시되는데, 라벨 스무딩을 적용 하면 0과 1 사이의 값으로 라벨을 부드럽게 만들어 줍니다.

### 라벨 스무딩 장점

+ 모델의 일반화 능력 향상 : 라벨 스무딩은 모델에 일부 허용 가능한 오차를 주어 일반화 능력을 향상 시킵니다. 이는 모델이 훈련 데이터에 너무 과도하게 적합되는 것을 방지하고, 실제 상황에서의 성 능을 향상시킬 수 있습니다.
<br></br>
+ 모델의 불확실성 학습: 라벨 스무딩은 모델에게 라벨이 절대적인 확실한 정보가 아니라는 것을 알려줍니다. 이는 모델이 불확실한 예측에 대해서도 적절하게 대응하도록 유도합니다.
<br></br>
+ 과적합 완화 : 라벨 스무딩은 모델이 훈련 데이터에 너무 맞추는 경향을 완화시킵니다. 이는 모델이 더 일반화된 패턴을 학습하도록 도와줍니다.

### 라벨 스무딩 한계

**라벨 스무딩은 과적합을 완화하고 모델의 성능을 향상시킬 수 있지만, 일부 상황에서는 한계가 있을 수 있습니다.**

+ 라벨 정보 손실 : 라벨 스무딩은 정확한 라벨을 부드럽게 만들어 불확실성을 반영하는 방식으로 동작합니다. 그러나 이는 라벨 정보의 일부 손실을 의미하기도 합니다. 따라서, 라벨 스무딩을 적용하면 모델이 정확한 라벨 정보에 대해 약간의 희생을 해야할 수있습니다.
<br></br>
+ 성능 평가 지표 변화 : 라벨 스무딩을 적용하면 모델의 예측이 더 다양해지고 확률값이 조정됩니다.이로인해 성능 평가 지표인 정확도 등이 변할 수 있습니다. 때로는 라벨 스무딩으로 인해 성능 평가 지표가 감소할 수도 있습니다.
<br></br>
+ 라벨 스무딩 파라미터 선택의 어려움 : 라벨 스무딩의 강도는 스무딩 파라미터로 제어됩니다. 하지만 이 파라미터를 적절히 선택하는 것은 어려운 일일 수 있습니다. 너무 낮은 값은 라벨 스무딩의 효과를 충분히 발휘 하지 못하게 하고, 너무 높은 값은 모델의 성능을 저하시킬 수 있습니다.
<br></br>
+ 클래스 불균형 처리의 한계 : 라벨 스무딩은 클래스 불균형 문제를 완전히 해결하지는 않습니다. 클래스 불균 형이 심각한 경우, 라벨 스무딩만으로는 적절한 해결책이 되지 않을 수 있습니다. 이런 경우에는 다른 클래스 불균형 처리 기법과 함께 사용하는 것이 필요할 수 있습니다.

## *라벨 스무딩 실습*

### *사용자 정의 손실함수 (라벨 스무딩 적용 크로스 엔트로피)*

```python
class LabelSmoothingLoss(nn.Module) : # 사용자 정의 손실 함수 
    def __init__(self, num_classes, smoothing) : 
        super(LabelSmoothingLoss, self).__init__()
        self.num_classes = num_classes  #-> 분류 문제 : 클래스 개수 
        self.smoothing = smoothing        #-> 라벨 적용되는 스무딩 정도 제어 값 -> 기본 0.0 스무딩 적용 x
        self.confidence = 1.0 - smoothing  #-> smothing 보정값 (정답에 대한 신뢰도 점수)
        
    def forward(self, pred, target) : # pred : 예측값 , target : 정답지 
        one_hot = torch.zeros_like(pred).scatter(1, target.unsqueeze(1),1) # 원-핫 인코딩 텐서 생성
        # target.unsqueeze(1)로 하여서 해당 라벨의 인덱스 정보를 가져온다. ex)[[0],[1],[2]]
        # 앞에 1은 특정 차원을 의미하고 뒤에 1을 해당 인덱스에 채워넣을 값을 설정 
        # -> 첫번째 1차원 0번째 자리에 1 -> 두 번재 1차원 1번째 자리에 1 -> ...
        
        
        smooth_label = one_hot * self.confidence + (1 - one_hot) * self.smoothing / (self.num_classes - 1)
        # 1-one_hot을 하면 0과 1이 바뀐다.
        # 원-핫 인코딩 텐서에서 라벨 스무딩을 적용 -> 신뢰도 점수
        loss = torch.sum(-smooth_label * torch.log_softmax(pred, dim=1), dim=1)
        # 스무딩 처리된 라벨과 예측 확률간 교차 엔트로피 사용해서 loss 계산 -> log_softmax 확률 과 라벨스무딩 요소 곱을 통해 수행 
        
        return torch.mean(loss) # 평균 손실 구해 주고 반환 
```

### *모델 정의*

```python
class MyModel(nn.Module) :
    def __init__(self) :
        super(MyModel, self).__init__()
        
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.relu1 = nn.ReLU(inplace=True)
        self.pool1 = nn.MaxPool2d(kernel_size=2)
        
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.relu2 = nn.ReLU(inplace=True)
        self.pool2 = nn.MaxPool2d(kernel_size=2)
        
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(64*5*5, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, 10)
        
    def forward(self, x) :
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.pool1(x)
        
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool2(x)
        
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu3(x)
        x = self.fc2(x)
        
        return x
```

### *데이터셋 로드 및 전처리*

```python
train_dataset = datasets.FashionMNIST(root="../data/", train=True, download=True, transform=transforms.ToTensor())
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
```

### *모델 선언*

```python
model = MyModel()
print(model)
```

### 옵티마이저 설정, loss 함수 설정

```python
# 옵티마이저 설정
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# 손실 함수정의
num_classes = 10

criterion = LabelSmoothingLoss(num_classes, smoothing=0.2)
no_criterion = LabelSmoothingLoss(num_classes, smoothing=0.0)
```

### 학습루프

```python
num_epochs = 20

train_losses_no_smoothing = []
train_losses_smoothing = []

for epoch in range(num_epochs) :
    train_loss_no_smoothing = 0.0
    train_loss_smoothing = 0.0
    
    for images, labels in train_dataloader :
        
        optimizer.zero_grad()
        
        # 라벨 스무딩 적용하지않는 경우
        outputs_no_smoothing = model(images)
        loss_no_smoothing = no_criterion(outputs_no_smoothing, labels)
        loss_no_smoothing.backward()
        optimizer.step()
        train_loss_no_smoothing += loss_no_smoothing.item()
        
        # 라벨 스무딩 적용하는 경우
        outputs_smoothing = model(images)
        loss_smoothing = criterion(outputs_smoothing, labels)
        loss_smoothing.backward()
        optimizer.step()
        train_loss_smoothing += loss_smoothing.item()
        
    # 에포크마다 손실 기록
    train_losses_no_smoothing.append(train_loss_no_smoothing / len(train_dataloader))
    train_losses_smoothing.append(train_loss_smoothing / len(train_dataloader))
```

```python
epochs = range(1, num_epochs +1)

plt.plot(epochs, train_losses_no_smoothing, label="No Smoothing")
plt.plot(epochs, train_losses_smoothing, label="Smoothing")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

<h2 style="color: green;">MixUp Training</h2>

**Data augmentation 기법의 일종으로 두 데이터와 라벨을 일정 비율로 섞어 새로운 데이터를 생성하는 방법**



### MixUp Training 장점

+ 일반화 성능 향상 : MixUp은 데이터를 조합하여 새로운 학습 샘플을 생성하는 방식이기 때문에 모델의 일반화 성능을 향상시킬 수 있습니다. 생성된 학습 샘플은 원본 데이터의 특성을 보존하면서 다양한 변형을 가지므로, 모델은 더 다양한 패턴과 특징을 학습할 수 있습니다.
<br></br>
+ 과적합 감소 : MixUp은 원본 데이터에 비해 더 일반적인 학습 샘플을 생성하므로 과적합을 감소시킬 수 있습니다. 생성된 학습 샘플은 원본 데이터의 특성을 희석시키는 효과가 있어 모델이 특정 데이터에 과도하게 적합되는 것을 방지할 수 있습니다.
<br></br>
+ 데이터 다양성 증가 : MixUp은 이미지와 라벨을 선형적으로 조합하여 새로운 데이터를 생성하는 방식이므로, 데이터의 다양성을 증가시킵니다. 이는 모델이 다양한 시나리오와 변형된 입력에 대해 더 강건한 학습을 수행할 수 있도록 도와줍니다.
<br></br>
+ 데이터 부족 상황에서 유용함 : 데이터 부족 상황에서 MixUp은 유용한 방법일 수 있습니다. 데이터를 조합하여 새로운 학습 샘플을 생성하므로, 제한된 데이터셋에서 더 많은 정보를 활용할 수 있습니다. 이는 데이터 부족으로 인한 과적합을 완화시키고 모델의 성능을 향상시킬 수 있습니다.
<br></br>
+ 적용이 비교적 간단함 : MixUp은 구현이 비교적 간단하고 적용하기 쉬운 방법입니다. 이미지와 라벨 의 선형 조합만으로 새로운 학습 샘플을 생성할 수 있으며, 기존의 데이터 파이프라인에 쉽게 통합할 수 있습니다.



## *MixUp Training 실습*

### *모델 정의*

```python
class MyModel(nn.Module) :
    def __init__(self) :
        super(MyModel, self).__init__()
        
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.relu = nn.ReLU()
        self.flatten = nn.Flatten()
        
        self.fc1 = nn.Linear(16 * 32 * 32, 128)
        self.fc2 = nn.Linear(128, 10)
        
    def forward(self, x) :
        
        x = self.conv1(x)
        x = self.relu(x)
        x = self.flatten(x)
        
        x = self.fc1(x)
        x = self.fc2(x)
        
        return x
```

### *데이터셋 정의*

```python
train_transform = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.AutoAugment(),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.3, 0.3, 0.3))
])

train_dataset = datasets.CIFAR10(root="../data/", download=True, transform = train_transform, train=True)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
```

```python
model = MyModel()

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

### *MixUp 함수 정의*

```python
def mixup_data(x, y, alpha=1.0) :
    batch_size = x.size(0)
    lam = torch.rand(batch_size, 1, 1, 1)  #0과 1사이이의 랜던한 값 lam -> mixup 비율
    lam = torch.max(lam, 1 - lam)  #최소값이 0.5이상으로 설정
    mixed_x = lam * x + (1-lam) * x.flip(dims=[0,2,3])
    """
    lam * x -> x 값에다가 lam 곱셈 
    x.flip(dims=[0,2,3]) : 배치 차원을 기준으로 뒤집기 -> (1-lam)비율을 곱셈
    """
    """
    0번째 축을 뒤집음으로써 배치의 순서가 역순으로 변경됩니다. 즉, 이미지 배치의 순서가 뒤집힙니다.
    2번째 축을 뒤집음으로써 이미지의 높이가 상하로 뒤집힙니다.
    3번째 축을 뒤집음으로써 이미지의 너비가 좌우로 뒤집힙니다.
    """
    
    indices = torch.randperm(batch_size)  #batch_size 크기만큼 랜덤한 정수 생성
    mixed_y = lam.squeeze() * y + (1-lam.squeeze()) * y[indices]
    # 랜덤한 정수비만큼 곱해서 섞어준다.
    """
    lam.squeeze() * y -> y라벨에다가 lam 비율값 곱해준다.
    (1 - lam.squeeze()) * y[indices] -> 랜덤한 라벨 y에 1 - lam.squeeze()(비율) 곱하기
    """
    # squeeze() 함수는 크기가 1인 차원을 제거하여 텐서의 차원을 압축합니다. -> 1차원으로 평탄화
    mixed_y = mixed_y.type(torch.long)
    
    return mixed_x, mixed_y
```

```python
def plot_images(images, labels, title) :
    fig, axes = plt.subplots(2, 5, figsize=(12,6))
    fig.suptitle(title, fontsize=16)
    labels = labels.numpy()
    for i, ax in enumerate(axes.flat) :
        image = images[i].squeeze()
        ax.imshow(image, cmap='gray')
        ax.set_title(f"Label:{labels[i]}")
        
    plt.show()
```

### *학습 loop 정의*

```python
num_epochs = 10

train_losses_no_mixup = []
train_losses_with_mixup = []

end_idx = 0
for epoch in range(num_epochs) :
    train_loss_no = 0.0
    train_loss = 0.0
    
    for inputs, labels, in train_loader :
        optimizer.zero_grad()
        
        images_temp, labels_temp = inputs, labels
        
        inputs, labels = mixup_data(inputs, labels)

        mixed_images = inputs.cpu().numpy()
        mixed_images = np.transpose(mixed_images, (0,2,3,1))
        mixed_images = np.squeeze(mixed_images)
        
        if end_idx == 0:
            plot_images(mixed_images, labels.squeeze(), "Mixed Image with label smoothing")
            end_idx = 1
            
        outputs_no_mixup = model(images_temp)
        outputs_mixup = model(inputs)
        
        labels = torch.squeeze(labels)  #convert lables to 1D tensor
        
        loss_no_mixup = criterion(outputs_no_mixup, labels)
        loss_mixup = criterion(outputs_mixup, labels)
        
        loss_no_mixup.backward()
        loss_mixup.backward()
        
        optimizer.step()
        
        # 배치 사이즈가 32라고하면 for문 한번에 32번의 loss값이 더해진다. 
        # -> train_loader가 끝날 때까지 for문 회귀(1 epoch 실행 loss값이 변수에 저장) 
        train_loss_no += loss_no_mixup.item()
        train_loss += loss_mixup.item()
        
    # 에포크 마다 손실 기록
    # 10번에 에포크마다 각 에포크에 더해진 loss값을 전체 데이터 길이로 나눈뒤에 배열에 추가
    train_losses_no_mixup.append(train_loss_no / len(train_loader))
    train_losses_with_mixup.append(train_loss / len(train_loader))
    # len(train_loader)는 훈련 데이터셋을 배치 단위로 나눈 경우의 배치 개수를 나타냅니다. 
    # 이 값은 전체 훈련 데이터의 개수를 배치 크기로 나눈 값입니다.
    # 결국 한 에포크마다 발생한 손실의 평균값을 의미
```

```python
epochs = range(1, num_epochs +1 )

plt.plot(epochs, train_losses_no_mixup, label='No Mixed')
plt.plot(epochs, train_losses_with_mixup, label='Mixed')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```
