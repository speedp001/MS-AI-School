<h1 style="color: blue;">가. 인공신경망 이론</h1>

<h2 style="color: green;">인공신경망 기본 개념</h2>

인공신경망(Artificial Neural Network, ANN)은 뇌의 신경망을 모방하여 만든 기계 학습 알고리즘입니다. 이는 인간 뇌에서 발견되는 다양한 뉴런들이 각각 정보를 처리하고 다른 뉴런들과 연결되어 높은 수준의 정보 처리를 가능케 하는 원리를 모방하고자 합니다.

**인공 신경망의 초기 단계에서 퍼셉트론이라는 단순한 인공 신경망 모델의 등장 -> XOR 비선형 문제 해결 불가 -> 역전파 알고리즘 등장**

<h2 style="color: green;">인공신경망의 구조</h2>

 인공 신경망은 입력층, 은닉층, 출력층으로 구성된 다층 퍼셉트론(MLP)을 중심으로 만들어집니다. 인공 신경망은 각 층의 노드들 사이에 가중치(weight)와 바이어스(bias)를 가집니다. 가중치는 입력값과 은닉층 노드, 은닉층 노드와 출력층 노드 간의 연결에 대한 정보를 담고 있습니다. 바이어스는 각 노드에 대한 편향 값을 의미합니다.

 신경망은 주어진 입력 데이터를 사용하여 각 층의 노드에서 계산된 값을 전달하면서 예측 값을 계산합니다. 이 때, 각 노드에서 계산된 값을 활성화 함수를 통해 변환합니다. 변환된 값은 다음 층의 노드로 전달되어 예측 값을 계산하는 과정을 반복하게 됩니다. 이 과정에서 신경망은 가중치와 바이어스를 조정하며 학습합니다.


## 입력층
입력층은 입력 데이터가 들어오는 부분으로, 보통은 데이터의 각 특성을 나타내는 노드들로 구성됩니다. 예를 들어, 손글씨 숫자 인식 문제에서는 이미지의 픽셀 값을 입력으로 사용할 수 있습니다.

## 은닉층
은닉층은 입력층과 출력층 사이에 있는 층으로, 신경망이 복잡한 패턴을 학습할 수 있도록 돕습니다. 여러 개의 은닉층을 가진 신경망도 있습니다. 은닉층의 노드는 활성화 함수를 사용하여 입력값을 계산합니다. 이 활성화 함수는 일반적으로 비선형 함수입니다. 비선형 함수를 사용하는 것은 신경망이 복잡한 함수를 모델링할 수 있도록 돕기 때문입니다.

## 출력층
출력층은 신경망이 예측하거나 분류해야 하는 출력 값을 내보내는 부분입니다. 출력층의 노드는 출력 값의 종류에 따라 달라집니다. 예를 들어, 이진 분류 문제에서는 하나의 출력 노드만 있으면 되지만, 다중 클래스 분류 문제에서는 여러 개의 출력 노드가 필요합니다.

## 가중치와 편향값


**가중치(weight)**는 인공신경망에서 사용되는 개념으로, 입력 신호에 곱해져서 다음 레이어로 전달되는 값입니다. 가중치는 학습 과정에서 자동으로 조정되어 최적의 값을 찾게 됩니다.

 가중치는 학습 과정에서 조정되어 최적의 값을 찾게되며, 이를 위해서 일반적으로 역전전(backpropagation)알고리즘이 사용됩니다. 역전파 알고리즘은 출력값과 실제값의 차이를 계산하여 각 노드의 가중치를 적절하게 수정하게 됩니다. 이러한 학습 과정을 통해 신경망은 입력 데이터를 정확하게 처리할 수 있는 최적의 가중치 값을 찾게 됩니다.

**편향(bias)은** 딥러닝에서 모델이 예측하는 값에 대한 상수항(constant term)입니다. 즉, 입력에 대한 가중치와 곱해지는 편향이 모델의 출력에 더해져 최종 예측 값을 만듭니다.

 딥러닝에서는 편향이 모델의 학습에 매우 중요합니다. 편향이 없으면 모델은 입력 데이터에 대한 직선만 학습할 수 있습니다. 즉, 입력 데이터와 출력 값 사이의 관계가 반드시 원점을 지나는 직선이어야 합니다. 하지만 대부분의 데이터는 이러한 직선과 원점을 지나지 않는 형태를 가지고 있기 때문에 모델이 적절한 예측을 할 수 없습니다.

 따라서, 편향을 추가함으로써 모델이 보다 복잡한 함수 형태를 학습할 수 있게 됩니다. 또한, 편향이 학습에 영향을 미치기 때문에 가중치와 마찬가지로 초기값을 적절하게 설정하는 것이 중요합니다.


## *인공신경망 가중치 적용한 시각화 실습*

```python
import numpy as np
```

```python
import torch
import matplotlib.pyplot as plt
```

```python
# 2차원 가중치 행렬을 랜덤으로 초기화
w = torch.randn(3,3)
print(w)

# 1차원으로 평탄화
w_flattend = w.flatten()
print(w_flattend)
```

<h2 style="color: green;">1. 손실 함수</h2>

 손실 함수(loss function)는 인공 신경망에서 학습 중 모델이 예측한 출력값과 실제 정답을 비교하여, 모델이 얼마나 잘 예측했는지를 측정하는 함수입니다. 손실 함수의 값은 모델이 예측한 값과 실제 값이 얼마나 차이가 나는지를 나타내며, 이 값을 최소화하는 방향으로 모델의 가중치를 조정하여 예측 성능을 개선합니다.
 
 손실 함수는 다양한 종류가 있으며, 문제의 종류나 출력값의 형태에 따라 적절한 손실 함수를 선택하여 사용합니다. 예를 들어, 회귀 문제에서는 평균 제곱 오차(Mean Squared Error, MSE) 손실 함수가 일반적으로 사용되며, 분류 문제에서는 크로스 엔트로피(Cross Entropy) 손실 함수가 일반적으로 사용됩니다.
 
 인공 신경망에서는 손실 함수를 최소화하는 방향으로 가중치를 업데이트하여 학습합니다. 이를 위해 역 전파 알고리즘이 사용되며, 손실 함수의 기울기(gradient)를 계산하여 가중치를 업데이트합니다.

## *평균 제곱 오차 실습 (손실 함수)*



 MSE는 이 실제값과 예측값의 차이를 제곱하여 평균을 구한 값으로, 예측값과 실제값의 차이가 클수록 오차가 커지는 특성을 가집니다. 따라서 모델이 예측한 값이 실제값과 가까울수록 MSE의 값은 작아지게 됩니다.

```python
def mean_squred_error(y_true, y_pred) :
    mean_temp = np.mean((y_true - y_pred)**2)
    
    return mean_temp

y_true = np.array([1,2,3,4])
y_pred = np.array([2,3,5,6])

mean_value = mean_squred_error(y_true, y_pred)
print(mean_value)
```

## *Binary Cross Entropy Loss 실습 (손실 함수)*

## *Cross Entropy Loss 실습 (손실 함수)*

**교차 엔트로피 손실**



Cross Entropy Loss는 예측값과 실제값 사이의 교차 엔트로피를 계산한 것입니다. 이진 분류나 다중 분류 문제에서 사용됩니다.

```python
def cross_entropy_loss(y_true_01, y_pred_01, c) :
    """
    y_true_01 -> 실제값(크기 : (n,c))
    y_pred_01 -> 예측값(크기 : (n,c))
    c         -> 클래스 개수
    """
    
    n_temp = y_true_01.shape[1]
    print("n_value", n_temp)
    ce_loss = -1 / n_temp * np.sum(y_true_01 * np.log(y_pred_01))
    
    return ce_loss

y_true_01 = np.array([[1,0,0], [0,1,0], [0,0,1]])  #실제값
y_pred_01 = np.array(([0.7, 0.2, 0.1], [0.1,0.8,0.1], [0.2,0.3,0.5]))  #예측값
c = 3

ce_loss_value = cross_entropy_loss(y_true_01, y_pred_01, c)
print(ce_loss_value)
```
