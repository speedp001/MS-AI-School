## 선형 분류 - 서포트 벡터 머신 (Support Vector Machine)

SVM은 Support Vector Machine의 약어로, 데이터를 분류하는 알고리즘 중 하나입니다. 기본 아이디어는 각 데이터 포인트들을 고차원 공간으로 매핑한 후, 그 공간에서 최대한 많은 데이터 포인트들을 서로 분리하는 초평면(hyperplane)을 찾는 것입니다. 이때, 분리된 두 클래스 사이의 거리(margin)를 최대화하는 초평면을 찾게 되며, 이를 위해 서포트 벡터(support vector)라는 개념을 사용합니다.

SVM은 1960년대부터 컴퓨터 비전, 패턴 인식, 자연어 처리 등 다양한 분야에서 활용되어 왔습니다. 하지만 초기 SVM 알고리즘은 컴퓨터 연산 능력이 한정되어 있던 시절에는 계산 복잡도가 매우 높아서 잘 사용되지 않았습니다. 1990년대에 들어서는 컴퓨터 연산 능력이 향상되면서, SVM 알고리즘도 발전하여 다양한 분야에서 활용되고 있습니다.

 SVM은 데이터를 분류하는 방법 중 하나로, **경계선으로부터 가장 멀리 떨어져 있는 데이터 포인트(서포 트 벡터)와 경계선 사이의 거리(margin)를 최대화하는 결정 경계를 찾는 방식입니다.** 이를 최대 마진 분류(maximum margin classification)이라고도 합니다.
 
 SVM은 두 개의 클래스를 구분하는 이진 분류(binary classification)에 주로 사용됩니다. 이를 위해서 **SVM 은 주어진 데이터를 고차원 공간(higher-dimensional space)으로 매핑하여 선형 분리가 가능하도록 합니다. 이때 매핑 함수를 커널(kernel)이라고 부릅니다.**
 
 SVM은 기존의 선형 분류 방식과 달리, **경계선과 가장 가까운 데이터 포인트(서포트 벡터)들**만을 사용하여 결정 경계를 찾기 때문에, 과적합(overfitting)을 방지하고 일반화 성능을 향상시킬 수 있습니다. SVM의 핵심 아이디어는 마진을 최대화하는 결정 경계를 찾는 것입니다.

### 서포트 벡터 머신(SVM) 장점

+ **이상치에 강합니다:** SVM은 라지 마진(large margin)을 생성하므로, 이상치에 대해서도 강건하게 분류할 수 있습니다.

    *라지마진 : 결정 경계(Decision Boundary)와 데이터 간의 거리(margin)를 최대화하는 것을 의미합니다. 즉, SVM은 데이터 간의 간격을 최대화하여 결정 경계를 찾는 것입니다. 이를 통해 SVM은 일반화 성능을 높이고, 이상치(outlier)에 덜 민감하도록 만들 수 있습니다. 이러한 라지 마진의 개념은 SVM의 핵심 아이디어 중 하나입니다.* <br></br>

+ **다양한 커널(Kernel)을 사용:** 커널 함수를 통해 고차원 데이터를 다룰 수 있으므로, 비선형 분류 문제도 해결할 수 있습니다.
<br></br>

+ **과적합(overfitting) 방지:** SVM은 라지 마진을 만드는 방식으로 모델을 학습하므로, 과적합 문제를 방지할 수 있습니다.
<br></br>

+ **높은 정확도:** SVM은 다른 분류 알고리즘보다 높은 정확도를 보이는 경우가 많습니다.
<br></br>

### 서포트 벡터 머신(SVM) 한계

+ **커널 함수 선택의 어려움:** 커널 함수의 선택은 매우 중요한데, 데이터의 성질과 맞지 않으면 성능이 낮아질 수 있습니다. 이를 조절하기 위해서는 매개변수 튜닝이 필요합니다.
<br></br>

+ **대용량 데이터에 대한 처리 어려움:** SVM은 데이터의 크기가 커지면 계산 복잡도가 증가하므로 대용량 데이터 처리에 어려움을 겪을 수 있습니다.
<br></br>

+ **다중 클래스 분류에 대한 처리 어려움:** SVM은 이진 분류(binary classification)에 대해서만 적용할 수 있으며 다중 클래스 분류(multi-class classification) 문제를 해결하기 위해서는 복잡한 전략이 필요합니다.

### 서포트 벡터 머신의 종류

+ **다중 클래스 SVM**

    SVM은 기본적으로 이진 분류 모델입니다. 다중 클래스 SVM은 이진 분류 모델을 여러 개 결합하여 다중 클래스 분류 문제를 해결하는 방법입니다. 일대다(one-vs.-rest) 방법과 일대일(one-vs.-one) 방법이 있습니다. <br></br>   

+ **회귀 SVM**

    SVM은 분류 뿐만 아니라 회귀 문제도 해결할 수 있습니다. 회귀 SVM은 입력 데이터의 값이 어떤 범위 안에 속하도록 하는 결정 경계를 찾습니다. 회귀 SVM은 일반적으로 제한된 범위 내에서 오차가 가장 작도록 하는 SVM을 이용합니다.

### SVM은 크게 선형 SVM과 비선형 SVM으로 구분

+ **선형 SVM**

    선형 SVM은 입력 데이터를 선형으로 분류하는 기능을 수행합니다. 이를 위해 하나의 초평면(hyperplane)을 찾아 데이터를 분류합니다. 선형 SVM은 데이터가 선형 분리 가능한 경우에 적용됩니다. <br></br>   

+ **비선형 SVM**

    비선형 SVM은 입력 데이터가 선형 분리 불가능한 경우 사용됩니다. 비선형 SVM은 커널 기법(kernel trick)을 이용해 고차원 공간으로 데이터를 변환하여 선형 SVM을 적용하는 것입니다. 대표적인 비선형 SVM의 종류로는 다항식 커널(Polynomial Kernel), 가우시안 커널(Gaussian Kernel), 시그모이드 커널(Sigmoid Kernel) 등이 있습니다.
    
### 선형 SVM

선형 SVM(Linear SVM)은 선형적으로 분리 가능한 데이터셋에서 사용하는 SVM입니다. 선형 SVM은 SVM의 가장 기본이 되는 형태로, 선형적인 결정 경계를 사용하여 데이터셋을 분류하는 방식입니다.

선형SVM의목적은마진을최대화하는결정경계를찾는것입니다.이를 위해 SVM은 각 클래스 사이의 가장 가까운 샘플로부터 일정한 거리(margin)를 유지하는 결정 경계를 찾습니다. 이때, 가장 가까운 샘플들을 서포트 벡터(Support Vector)라고 합니다.

선형 SVM에서는 선형 결정 경계를 찾기 위해 선형 커널을 사용합니다. 선형 커널은 입력 데이터의 내적(in ner product)을 계산하여 새로운 특성 공간을 만들어내는데, 이를 통해 비선형 결정 경계를 선형 결정 경계로 매핑할 수 있습니다.

## *PyTorch를 사용한 선형 SVM 모델 학습 및 결정 경계 시각화*

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
# 데이터 생성
x = torch.randn(100, 2)
# 크기가 (2500, 2)인 랜덤한 값을 가지는 텐서 x를 생성
y = torch.where(x[:,0] > x[:,1], 1, -1).float()
# x 텐서의 첫 번째 열과 두 번째 열의 값을 비교하여 조건을 만족하는 경우에는 1, 그렇지 않은 경우에는 -1

# x, y로 데이터셋 생성
dataset = TensorDataset(x, y)
loader = DataLoader(dataset, batch_size=36)
```

```python
# SVM 모델 정의
class SVM(nn.Module) :
    def __init__(self) :
        super().__init__()
        self.linear = nn.Linear(2,1)
        
    def forward(self, x) :
        return self.linear(x)
```

```python
# 모델, 손실 함수, 옵티마이저 생성
model = SVM()
criterion = nn.HingeEmbeddingLoss() # 양성 1 음성 -1 클래스로 지정되어 있어야 사용가능
# 주어진 입력과 대상 사이의 경계선 손실값 계산하는 함수
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# 모델 학습
for epoch in range(100) :
    for inputs, targets in loader :
        output = model(inputs)
        loss = criterion(output.squeeze(), targets)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    if (epoch+1) % 10 == 0 :
        print(f"Epoch [{epoch+1} / 100], Loss : {loss.item():4f}")
```

```python
# 시각화
import matplotlib.pyplot as plt
import numpy as np
 
w = list(model.parameters())[0][0]
b = list(model.parameters())[1].item()
slope = (-w[0] / w[1]).detach().numpy()
intercept = (-b / w[1]).detach().numpy()
x_range = np.linspace(-3, 3, 100)
y_range = slope * x_range + intercept

plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.plot(x_range, y_range, 'k-')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```

### 비선형 SVM

커널 SVM은 SVM(Support Vector Machine) 알고리즘의 한 종류로, 선형으로 분리되지 않는 비선형 데이터를 분류하는 데 사용됩니다.

기본적인 SVM 알고리즘에서는 초평면(hyperplane)을 이용하여 데이터를 분류하는데, 이 때 초평면의 결정 경계는 선형 함수로 표현됩니다. 하지만 실제 데이터는 선형으로 분리되지 않을 때가 많기 때문에, 이를 해결하기 위해 비선형 변환을 통해 데이터를 고차원 공간으로 매핑한 후, 선형분류를 수행하는 방법이 커널 SVM입니다.

커널 SVM은 비선형 변환으로 인한 고차원 공간에서의 계산 비용 문제를 해결하기 위해, 커널(Kernel) 함수를 사용합니다. 커널 함수는 두 데이터 간의 유사도를 계산하여 고차원 공간에서의 내적 연산을 대신합니다. 이를 통해 고차원 공간에서의 분류 문제를 해결할 수 있습니다.

커널 SVM의 핵심은 서포트 벡터(Support Vector)입니다. 서포트 벡터란, 결정 경계와 가장 가까이 위치한 데이터 포인트를 말합니다. SVM은 이 서포트 벡터와 초평면 사이의 거리(margin)를 최대화하는 결정 경계를 찾는 것이 목적입니다. 이 때 최대 마진 결정 경계를 찾는 것이 최적화 문제이며,이를 위해 이차 계획법(Quadratic Programming)을 사용합니다.

커널 SVM은 RBF(Radial Basis Kernel) 커널, 다항 커널(Polynomial Kernel), 다양한 커널(Gaussian Kernel), Gaussian Kernel)가 있다. 적절한 커널 함수를 선택하는 것이 분류 성능에 큰 영향을 미칩니다. 커널 함수의 선택은 trial and error를 통해 적절한 값을 찾는 것이 일반적입니다.

## *One_class SVM을 사용하여 데이터 이상 탐지*

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.svm import OneClassSVM
```

```python
x, y = make_moons(n_samples=1000, noise=0.1, random_state=777)

# 데이터 시각화
plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.coolwarm)
plt.show()

# 이상 데이터 설정
outliers_ratio = 0.1  # 이상치 비율
n_outliers = int(outliers_ratio * len(x))  # 이상치 데이터 개수
n_inliers = len(x) - n_outliers  # 정상 데이터 개수

rng = np.random.RandomState(777)  # 난수 생성 객체
x_outliers = 1.5 * rng.uniform(low=-1, high=3, size=(n_outliers, 2))  
# 이상치 데이터 생성 난수 생성을 이용해서 uniform 분포를 따르는 값을 생성 -> 1.5배 해서 이상치 데이터를 생성
x = np.vstack([x, x_outliers])
# 이상치 데이터를 기존 데이터에 추가
y = np.hstack([y, -1 * np.ones(n_outliers)])
# 이상치 데이터의 클래스는 -1로 설정
```

```python
# 데이터 정규화
x = (x - x.mean(axis=0)) / x.std(axis=0)

# 학습 데이터 평가 데이터 분리
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1, random_state=777)

# One_class SVM 모델 학습
model = OneClassSVM(nu=0.05, kernel='rbf', gamma='auto')
model.fit(x_train)

# 이상치 탐지
y_pred = model.predict(x_test)
y_pred = np.where(y_pred == 1, 0, y_pred) #1을 0으로 -1을 1로 변환
score = f1_score(y_test, y_pred, average='macro')

"""
F1 score는 정밀도와 재현율의 조화 평균으로 계산 됩니다.

정밀도 = TP / (FP + TP)
재현율 = TP / (TP + FN)

F1 score = 2 * (정밀도 * 재현율) / (정밀도 + 재현율)

FN : 거짓 음성 -> 실제로 양성인데 음성이라고 하는 경우
TP : 참 양성 -> 실제 양성인데 양성이라고 하는 경우
FP : 거짓 양성 -> 실제는 음성인데 양성이라고 하는 경우

F1 스코어 : 0부터 1까지 값을 가지고 1에 가까우면 좋은 모델
"""

print(f"F1 score >> ", score)
```

```python
# 모든 데이터 포인트에 대한 예측 결과 얻기
y_pred_all = model.predict(x)
 
# 이상치 데이터와 정상 데이터를 분리하여 시각화
X_inliers = x[y == 0]
X_outliers = x[y == -1]
y_pred_inliers = y_pred_all[y == 0]
y_pred_outliers = y_pred_all[y == -1]
 
plt.scatter(X_inliers[:, 0], X_inliers[:, 1], c=y_pred_inliers, cmap=plt.cm.coolwarm)
plt.scatter(X_outliers[:, 0], X_outliers[:, 1], c=y_pred_outliers, cmap=plt.cm.coolwarm, marker='x', 
            s=100, linewidth=2)
plt.show()
```

## *SVM을 활용한 이미지 분류*

```python
import time
import torch
import torchvision
import torchvision.transforms as transforms
from sklearn.svm import SVC
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# 데이터 로드
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5), (0.3))
])

train_dataset = torchvision.datasets.MNIST(root="../data", train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root="../data", train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=1000, shuffle=True, num_workers=2)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=1000, shuffle=False, num_workers=2)
```

```python
# 특징 추출
train_features = []
train_labels = []

for i, data in enumerate(train_loader, 0) :
    # 뒤에 인자값 0은 시작점을 의미
    inputs, labels = data
    # print(inputs.shape)
    inputs = inputs.view(inputs.shape[0], -1)
    # inputs 텐서의 크기는 (1000, 1*28*28)인 2차원 텐서가 됩니다.
    
    train_features.append(inputs)
    train_labels.append(labels)

torch.Size([1000, 1, 28, 28])
train_features = torch.cat(train_features, dim=0)
train_labels = torch.cat(train_labels, dim=0)

test_features = []
test_labels = []

for i, data in enumerate(test_loader, 0) :
    # 뒤에 인자값 0은 시작점을 의미
    inputs, labels = data
    # print(inputs.shape)
    inputs = inputs.view(inputs.shape[0], -1)
    test_features.append(inputs)
    test_labels.append(labels)

torch.Size([1000, 1, 28, 28])
test_features = torch.cat(test_features, dim=0)
test_labels = torch.cat(test_labels, dim=0)
```

```python
# SVM을 활용한 이미지 분류
print("SVM MODEL TRAIN AND TEST")
start_time = time.time()
model = SVC(C=1.0, kernel='rbf', gamma=0.01)

# sklearn 학습 모델이므로 numpy()형식으로 넣어주어야한다.
model.fit(train_features.numpy(), train_labels.numpy())
acc = model.score(test_features.numpy(), test_labels.numpy())

end_time = time.time()
print("ACC >> ", acc)
train_time = end_time - start_time
print("Training Time : {:.2f}seconds".format(train_time))
```

```python
# 예측 결과 시각화
def plot_images(images, labels, preds):
    fig, axes = plt.subplots(1, 10, figsize=(10, 1))

    for i, ax in enumerate(axes):
        ax.imshow(images[i].reshape((28, 28)), cmap='gray')
        title = 'label: {}\npred: {}'.format(labels[i], preds[i])
        ax.set_title(title)
        ax.axis('off')
    plt.subplots_adjust(top=0.5, bottom=0, hspace=0, wspace=0.5)
    plt.show()

# 테스트 이미지 가져오기
test_images, test_labels = next(iter(test_loader))
test_images = test_images.view(test_images.shape[0], -1)

# 모델 예측
test_preds = model.predict(test_images.numpy())

# 결과 시각화
plot_images(test_images.numpy(), test_labels.numpy(), test_preds)
```

### SVM의 하이퍼파라미터 최적화 - 그리드 서치 소개

SVM 모델에서 사용하는 하이퍼파라미터들은 모델의 성능과 결과에 큰 영향을 미칩니다. 하이퍼파라미터를 잘 선택하면 모델의 성능을 높일 수 있지만, 이를 선택하는 것은 어려운 일입니다. 이 때문에 SVM 모델에서는 하이퍼파라미터를 선택하는 과정을 자동화하는 방법인 그리드 서치(Grid Search)를 사용합니다.

그리드 서치는 SVM 모델에서 사용하는 하이퍼파라미터들의 모든 조합을 시도하며, 최적의 조합을 찾아내는 방법입니다. 예를 들어, SVM 모델에서 C 값이 1, 10, 100이 가능하고, kernel 값이 'linear', 'rbf'가 가능 하다고 가정하면, 이들을 모두 조합하여 6가지 경우를 시도합니다. 그리고 이 중 가장 좋은 성능을 보이는 조합을 선택합니다.

그리드 서치를 수행할 때는 모든 하이퍼파라미터 조합을 시도하기 때문에 시간이 많이 소요됩니다. 따라서 그리드 서치는 하이퍼파라미터 공간을 충분히 탐색하면서도 연산 비용을 최소화하는 것이 중 요합니다. 이를 위해 scikit-learn 패키지에서는 GridSearchCV 클래스를 제공합니다. 이 클래스를 사용 하면 하이퍼파라미터 공간을 적절히 설정하고, 교차 검증(Cross Validation)을 사용하여 모델의 성능을 측정하고, 최적의 하이퍼파라미터 조합을 찾아낼 수 있습니다.

## *GridSearchCV SVM 모델 최적 하이퍼파라미터 찾기*

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
# 데이터 생성
rng = np.random.RandomState(0)
x = rng.randn(100, 10)
y = rng.binomial(1, 0.5, 100)
```

```python
# 파이프라인 생성
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC())
])
```

```python
# 탐색할 하이퍼파라미터 그리드 생성
param_grid = {
    'svm__C' : [0.1, 1, 10],
    'svm__kernel' : ['linear', 'rbf'],
    'svm__gamma' : ['scale', 'auto']
}
```

```python
# 그리드 서치 수행
grid_search = GridSearchCV(svm_pipeline, param_grid, cv=5)
grid_search.fit(x,y)
```

```python
# 최적 하이퍼 파라미터와 최고 성능 출력
print("Best parameters : ", grid_search.best_params_)
print("Best cross-validation score : ", grid_search.best_score_)
```

### SVM의 하이퍼파라미터 최적화 - 랜덤 서치 소개

랜덤 서치는 하이퍼파라미터의 값들을 랜덤으로 선택하여 모델을 학습시키고, 그 결과를 바탕으로 하이퍼 파라미터의 최적 조합을 찾는 방법입니다.

랜덤 서치를 사용하면 하이퍼파라미터의 모든 조합을 시도해보는 것보다 효율적으로 하이퍼파라미터를 탐색할 수 있습니다. 하지만 랜덤 서치 역시 하이퍼파라미터의 범위를 어떻게 설정하느냐에 따라서 최적의 조합을 찾는데 실패할 수도 있습니다.

랜덤 서치를 구현하기 위해서는 우선 SVM 모델을 정의하고, 모델을 학습시키는 함수를 작성해야 합니다. 그리고 랜덤으로 선택할 하이퍼파라미터의 범위를 설정하고, 이를 바탕으로 랜덤하게 하이퍼파라미터를 선택하여 SVM 모델을 학습시키고, 검증 데이터를 바탕으로 모델의 성능을 평가합니다. 이를 여러 번 반복하여 최적의 하이퍼파라미터 조합을 찾습니다.

## *SVM 하이퍼파라미터 최적화 - 랜덤 서치*

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from scipy.stats import loguniform

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
# 데이터 로드
digits = load_digits()

# 독립 변수, 종속 변수 데이터 지정
x = digits.data
y = digits.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# 탐색할 하이퍼파라미터 공간
params = {
    "C" : loguniform(1e-4, 100),
    "kernel" : ["linear", "poly", "rbf", "sigmoid"],
    "gamma" : ['scale', 'auto'] + list(loguniform(1e-4, 10).rvs(10)),
    "degree" : range(1,6),
    "coef0" : loguniform(1e-4, 10).rvs(10)
}
```

```python
svm_model = SVC()
random_search = RandomizedSearchCV(svm_model, params, n_iter=100, cv=5, verbose=2, n_jobs=-1)
random_search.fit(x_train, y_train)
print(random_search.best_params_)

# 최적 모델 저장
best_model = random_search.best_estimator_

# 테스트 데이터에 대한 예측 수행
y_pred = best_model.predict(x_test)

print("ACC > ", accuracy_score(y_test, y_pred))
```

### SVM의 하이퍼파라미터 최적화 - 베이즈 최적화

 베이즈 최적화는 머신러닝 모델의 하이퍼파라미터 최적화에 사용되는 알고리즘 중 하나입니다. 베이즈 최적화는 목적 함수를 평가하고, 모델의 하이퍼파라미터를 조정하며, 평가 결과를 사용하여 목적 함수를 효율적으로 탐색하는 방법입니다.

 베이즈 최적화는 목적 함수의 값을 최소화하거나 최대화하는 하이퍼파라미터 조합을 찾기 위해 베이즈 정리를 사용합니다. 이를 위해, 이전에 시도한 하이퍼파라미터 조합에 대한 정보와 함께 현재 상태의 목적 함수 값을 기반으로 하이퍼파라미터 조합을 선택합니다. 이전 시도에서 획득한 정보는 사전 확률 분포로 사용됩니다.

 베이즈 최적화는 전역 최적점을 찾을 가능성이 높으며, 상대적으로 작은 횟수로 최적점을 찾을 수 있습니다. 하지만, 일반적으로 랜덤 서치나 그리드 서치보다 계산 비용이 높습니다.
 
### 다중 클래스 SVM

다중 클래스 SVM은 세 개 이상의 클래스를 분류하는 문제에 사용되는 SVM입니다. 이전에 설명한 이진 분류 SVM과 달리, 다중 클래스 SVM은 분류할 클래스 수에 따라 여러 개의 결정 경계를 생성합니다. 다중 클래스 SVM은 일대다(OvA) 접근 방식 또는 일대일(OvO) 접근 방식으로 작동할 수 있습니다.

OvA 접근 방식은 각 클래스와 나머지 모든 클래스를 비교하여 이진 분류 문제로 바꿉니다. 그런다음, 모든 이진 분류기 중에서 가장 높은 점수를 받은 클래스를 선택합니다. OvO 접근 방식은 각 클래스 조합에 대해 이진 분류 문제를 만듭니다. 따라서 N개의 클래스가 있으면 N(N-1)/2개의 분류 문제를 생성합니다. 그런 다음, 각 분류기의 투표를 기반으로 가장 많은 투표를 받은 클래스를 선택합니다.

### SVM에서의 비대칭 데이터 문제 해결법

+ **데이터 샘플링:** 대다수 클래스의 샘플링 비율을 줄이거나 소수 클래스의 샘플링 비율을 늘립니다. 이를 통해 데이터를 균형있게 만들어줄 수 있습니다.
<br></br>
+ **비용 조절:** SVM은 분류 오류를 최소화하는 초평면을 찾는 것이 목적입니다. 비대칭 데이터에서는 소수 클래스를 잘못 분류하는 것이 더 큰 문제가 되므로, 소수 클래스의 분류 오류를 더 큰 비용으로 부과합니다. 이를 통해 SVM이 소수 클래스를 더욱 중요하게 취급하게 할 수 있습니다.
<br></br>
+ **커널선택:** SVM은 커널 함수를 사용하여 고차원공간에서 분리 초평면을 찾습니다. 커널 함수를 선택하는 것은 SVM의 성능에 큰 영향을 미칩니다. 비대칭 데이터에서는 일반적으로 비선형 커널 함수를 사용하여 더 복잡한 결정 경계를 생성합니다.
<br></br>
+ **클래스 가중치:** SVM에서는 클래스에 가중치를 부여하여 분류를 수행합니다. 이를 통해 소수 클래스를 더 중요하게 취급할 수 있습니다.
