<h2 style="color: green;">선형 회귀 (Linear Regression)</h2>

선형 회귀(Linear Regression)는 종속 변수와 한 개 이상의 독립 변수 간의 관계를 모델링하는 데 사용되는 선형 접근 방법입니다. 이 방법은 종속 변수와 독립 변수 사이에 선형적인 관계가 있다고 가정합니다. 선형회귀는 모델링, 예측 및 인과관계 분석에 널리 사용됩니다.

선형회귀의 목적은 독립변수와 종속변수 사이의 관계를 모델링하고, 독립변수의 값을 알 때 종속변수를 예측하는 것입니다. 이를 위해 데이터 셋을 사용하여 모델 파라미터를 학습하고, 새로운 독립 변수 값을 사용하여 종속 변수를 예측할 수 있습니다.

선형 회귀는 지도 학습(Supervised Learning)의 한 종류로, 회귀(Regression) 문제에 적용됩니다. 회귀 문제란, 연속적인 값(실수)을 예측하는 문제를 말합니다.

## *단순 선형 회귀 실습*

+ 단순 선형 회귀는 독립변수 x와 이에 대응하는 종속변수 y간의 선형적 관계를 모델링하는 회귀 분석 기법입니다. 다른 말로 하면, 주어진 데이터에 가장 적합한 직선을 찾는 것입니다. 이때, x가 주어졌을 때 y값을 예측할 수 있습니다.

+ 단순 선형 회귀에서는 x와 y간의 선형 관계를 수식으로 나타내어야 합니다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 예시 데이터 생성
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])
print(x)
print(y)

# 기울기와 절편 구하기 (기울기와 절편을 구하는 공식 적용)
beta_1 = np.sum((x-np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x))**2)
beat_0 = np.mean(y) - beta_1 * np.mean(x)

# 결과 출력
print("기울기 : ", beta_1)
print("절편 : ", beat_0)
```

## *최소 제곱법을 이용한 단순 선형 회귀 실습*

```python
# 데이터 생성
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])

# 산점도 그리기
plt.scatter(x,y)
plt.show()
```

```python
# 회귀선 그리기
plt.scatter(x,y)
plt.plot(x, beat_0 + beta_1 * x, color='red')
plt.show()
```

<h2 style="color: green;">다중 선형 회귀</h2>

다중 선형 회귀(Multiple Linear Regression)는 단순 선형 회귀와 마찬가지로 종속 변수와 독립 변수 간의 선형 관계를 모델링하는 회귀 분석 기법입니다. 다만, 독립 변수의 개수가 1개가 아닌 여러 개인 경우를 다룹니다.

다중 선형 회귀에서는 독립 변수의 개수를 p라고 표기하며, 종속 변수 y와 독립 변수 x_1, x_2, ..., x_p간의 선형 관계를 다음과 같은 수식으로 표현할 수 있습니다.

y = beta_0 + beta_1 x_1 + beta_2 x_2 + ... + beta_p x_p

<h2 style="color: green;">선형 회귀 평가 지표 소개</h2>

개념 : 회귀(Regression) 모델은 입력 변수(X)와 출력 변수(y) 사이의 관계를 모델링하는 머신 러닝 기술입니다. 이 모델의 목적은 주어진 입력 값에 대해 출력 값을 예측하는 것입니다.

**회귀 모델의 성능을 평가하기 위해 일반적으로 사용되는 지표는 다음과 같습니다.**

+ MSE(Mean Squared Error) : 예측 값과 실제 값 간의 차이를 제곱하여 평균을 구한 값입니다. 예측 값과 실제 값의 차이가 클수록 값이 더욱 커집니다. MSE는 회귀 모델에서 가장 널리 사용되는 지표 중 하나입니다.
<br></br>
+ RMSE(Root Mean Squared Error) : MSE에 루트를 씌운 값입니다. MSE보다는 값이 작고, 이 값이 작을수록 모델의 성능이 더 좋다고 할 수 있습니다.
<br></br>
+ MAE(Mean Absolute Error) : 예측 값과 실제 값 간의 차이를 절대값으로 계산하여 평균을 구한 값입니다. MAE는 MSE와 달리 제곱을 하지 않으므로 오차가 클 때 패널티가 덜 적용됩니다.
<br></br>
+ R-squared : 모델이 주어진 데이터에 대해 얼마나 잘 적합한지를 나타내는 지표입니다. R-squared는 0과 1사이의 값을 가지며, 1에 가까울수록 모델이 데이터를 잘 설명하고 있음을 나타냅니다.

## *선형 회귀 모델의 실제 활용 사례*

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_california_housing
```

```python
california_housing = fetch_california_housing()
df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
df['target'] = california_housing.target
df
```

```python
# 주택  가격과 다른 변수들 간의 상관관계 시각화
selected_columns = ['MedInc', 'AveOccup', 'AveRooms', 'AveBedrms']
pd.plotting.scatter_matrix(df[['target'] + selected_columns], figsize=(10, 10), diagonal='hist')
plt.show()
```

```python
california_housing = fetch_california_housing()
x, y = california_housing.data, california_housing.target
# x에는 캘리포니아 집 요소 데이터들이 들어있고 y에는 캘리포니아 집 값 정보가 들어있다.
# california_housing.data: 특징 행렬, 각 주택의 특징을 나타냄.
# california_housing.target: 타겟 변수, 각 주택의 가격을 나타냄.
print(x.shape, y.shape)

# 표준화
scaler = StandardScaler()
x = scaler.fit_transform(x)

# 1차원 -> 2차원
y = y.reshape(-1, 1)
# print(y)

# 데이터셋 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
# 선형 회귀 모델 정의
import torch

class LinearRegression(torch.nn.Module) :
    def __init__(self) :
        super(LinearRegression, self).__init__()
        # x_train.shape -> (16512, 8)
        
        self.linear = torch.nn.Linear(x_train.shape[1], 1)
        # x_train.shape[1]은 두 번째 차원의 크기, 즉 특징의 개수가 됩니다.
        # 1은 출력이 1개인 선형 레이어를 의미합니다.
    
    def forward(self, x) :
        out = self.linear(x)
        return out
```

```python
# 모델 학습
model = LinearRegression()

# loss 함수 지정
criterion = torch.nn.MSELoss()

# optimizer 지정
optimizer = torch.optim.SGD(model.parameters(), lr=0.00025)

for epoch in range(10000) :
    # NumPy 배열인 x_train을 PyTorch 텐서로 변환하는 역할
    inputs = torch.autograd.Variable(torch.from_numpy(x_train).float())
    labels = torch.autograd.Variable(torch.from_numpy(y_train).float())
    
    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()
    
    if epoch % 1000 == 0 :
        print("epoch {}, loss {}".format(epoch, loss.item()))
```

```python
# 모델 평가
model.eval()  # 평가모드로 진행
x_test = torch.autograd.Variable(torch.FloatTensor(x_test))
y_test = torch.autograd.Variable(torch.FloatTensor(y_test))

y_pred = model(x_test)

mse = criterion(y_pred, y_test)
rmse = np.sqrt(mse.item())
print("RSME >> ", rmse)
```

```python
# 실제 집값과 예측한 집값 차이 비교
with torch.no_grad() :  # 자동 미분 방지 -> 비활성화 -> 기울기를 구할 필요 없음
    x_sample = x_test[0]
    y_true = y_test[0].item()
    y_pred = model(x_sample).item()
    
    print(f"실제 집값 : {y_true:.2f}, 예측 집값 : {y_pred:.2f}")
```
