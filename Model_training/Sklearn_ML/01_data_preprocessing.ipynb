### 먼저 머신러닝의 대표적인 데이터 셋 중 하나인 전복 데이터를 불러오겠습니다. 전복 데이터셋은 수컷, 암컷, 유아기 3개의 범부로 이루어진 범주형 변수와 같이, 직경, 높이, 무게 등 여러 수치형 변수로 이루어져 있습니다. 데이터를 불러온 후 입력으로 사용할 변수들과 레이블로 사용할 성별 변수로 나누겠습니다.

```python
abalone_path = join('.', 'abalone.txt')
column_path = join('.', 'abalone_attributes.txt')
#.은 디렉토리를 나타낸다.

abalone_columns = list()
for l in open(column_path):
  abalone_columns.append(l.strip())
  #strip()을 이용해서 문자생략(공백이나 \n, \t)
#print(abalone_columns)

#Pandas를 이용해서 abalone 데이터를 읽어온다.
data = pd.read_csv(abalone_path, header=None, names=abalone_columns)
data
```

```python
#데이터 전처리 모듈의 불러오기 및 정의하기
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler()
mMscaler.fit(data)
data3 = mMscaler.transform(data)
#numpy형태로 나온다
data3
```

```python
from sklearn.preprocessing import StandardScaler

sdscaler = StandardScaler()
sdscaled_data = sdscaler.fit_transform(data)
sdscaled_data = pd.DataFrame(sdscaled_data, columns=data.columns)
sdscaled_data.head()
```

```python
#Sampling
#1.Random over, Under Sampling

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

ros = RandomOverSampler()
rus = RandomUnderSampler()
```

```python
#Over Sampling
oversampled_data, oversampled_label = ros.fit_resample(data, label)
oversampled_data = pd.DataFrame(oversampled_data, columns=data.columns)
```

```python
print('원본 데이터의 클래스 비율 \n{0}'.format(pd.get_dummies(label).sum()))
```

```python
#한쪽 데이터만 커지면 영향력이 골고루 분포되지 않으므로 데이터분석에 차질이 생긴다.
#oversampling을 이용하여 데이터가 적을쪽을 랜덤으로 데이터를 복제하며 데이터량을 늘린다.

print('\nRandom Over 샘플링 결과 \n{0}'.format(pd.get_dummies(oversampled_label).sum()))
```

```python
#Under Sampling
undersampled_data, undersampled_label = rus.fit_resample(data, label)
undersampled_data = pd.DataFrame(undersampled_data, columns=data.columns)
```

```python
print('\nRandom Over 샘플링 결과 \n{0}'.format(pd.get_dummies(undersampled_label).sum()))
#작은 쪽으로 수가 맞춰진다.
```

```python
#하지만 이렇게 많은 데이터량을 복제하다보면 데이터편향이 생긴다.
#SMOTE를 통해 해결
```

```python
#SMOTE

from imblearn.over_sampling import SMOTE
smote = SMOTE(k_neighbors=5)
```

```python
from sklearn.datasets import make_classification
data, label = make_classification(n_samples=1000, 
                                  n_features=2, 
                                  n_informative=2,
                                  n_redundant=0,
                                  n_classes=3,
                                  n_clusters_per_class=1,
                                  weights=[0.05, 0.15, 0.8],
                                  class_sep=0.8,
                                  random_state=2019
                                  )
#n_samples는 표본 데이터의수, n_features는 독립 변수의 수, n_classes 종속 변수의 클래스 수
#n_clusters_per_class=1 클래스별로 집합이 생긴다. weight는 가중치
```

```python
label
```

```python
fig = plt.Figure(figsize=(12,6))
plt.scatter(data[:,0],data[:,1], c=label, linewidth=1, edgecolor='k')
```

```python
smoted_data, smoted_label = smote.fit_resample(data, label)
```

```python
print('원본 데이터의 클래스 비율\n{0}'.format(pd.get_dummies(label).sum()))
print('SMOTE 데이터의 클래스 비율\n{0}'.format(pd.get_dummies(smoted_label).sum()))
```

```python
fig = plt.Figure(figsize=(12,6))
plt.scatter(smoted_data[:,0], smoted_data[:,1], c=smoted_label, linewidth=1, edgecolor='k')
```

```python
#One-hot Encodeing
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
```

```python
label.values.reshape((-1,1))  #1개씩 넣어서 만들어줘
ohe.fit(label.values.reshape((-1,1)))
one_hot_encoded = ohe.transform(label.values.reshape((-1,1)))
```

```python
columns = np.concatenate([np.array(['label']),ohe.categories_[0]])
columns
```

```python
result = pd.DataFrame(data = np.concatenate(
                      [label.values.reshape((-1,1)), one_hot_encoded.reshape((-1, 3))], 
                      axis=1),columns=columns)
result.head(10)
```
