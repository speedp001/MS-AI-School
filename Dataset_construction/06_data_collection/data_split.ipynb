<h1 style="color: blue;">마. 데이터 세트 분리</h1>



인공 신경망(Neural Network)에 사용되는 데이터는 다음과 같습니다. 
1. Full Dataset (Original Data Set) 
2. 훈련용 데이터 (Training Data), 검증용 데이터 (Validation Data), 테스트 데이터 (Testing Data) 로 분류 됩니다.

<h2 style="color: green;">훈련용 데이터(Training Data)</h2>

Training Data는 기계 학습에서 모델을 학습시키기 위해 사용되는 데이터 셋입니다. 모델이 정확하게 예측할 수 있는 대규모의 데이터 세트가 필요합니다. 이 데이터 셋은 모델의 학습을 개선하기 위한 다양한 기술들을 적용할 때 사용됩니다.

훈련 데이터는 일반적으로 레이블링 된 데이터가 포함됩니다. 레이블링 된 데이터는 이미지, 음성 또는 텍스트와 같은 입력 데이터와 해당 데이터의 분류 또는 예측에 대한 정답 레이블로 구성됩니다.

이를 통해 모델은 입력 데이터와 정답 레이블 사이의 연관성을 학습하여 새로운 입력에 대한 예측을 수행할 수 있습니다.

<h2 style="color: green;">검증용 데이터(Validation Data)</h2>

검증 데이터(validation data)는 모델 학습 과정에서 모델의 일반화 성능을 평가하기 위해 사용되는 데이터입니다. 일반적으로 학습 데이터를 이용하여 모델을 학습시킨 후, 검증 데이터를 이용하여 학습된 모델의 성능을 평가하고, 이를 바탕으로 모델의 하이퍼파라미터를 조정하거나 모델 아키텍처를 개선합니다.

검증 데이터는 학습 데이터와 동일한 분포에서 추출된 샘플로 이루어져 있으며, 학습 데이터와 달리 모델의 파라미터를 업데이트하는 데에는 사용되지 않습니다. 

대신 검증 데이터를 이용하여 모델의 성능을 평가합니다. 이를 통해 모델이 학습 데이터에 과적합(overfitting)되는 것을 방지하고, 실제 데이터에서의 일반화 성능을 높이는 것을 목표로 합니다.

검증 데이터는 보통 학습 데이터의 일정 부분을 임의로 분리하여 사용하며, 분리된 검증 데이터는 모델 학습이 끝난 후에도 최종 성능 평가를 위해 사용됩니다.

<h2 style="color: green;">테스트 데이터(Testing Data)</h2>

테스트 데이터는 모델의 일반화 성능을 평가하기 위해 사용됩니다. 모델이 학습한 데이터와는 완전히 독립적인 새로운 데이터로 구성되어 있으며, 모델이 이전에 본 적이 없는 데이터를 예측하도록 합니다. 

테스트 데이터는 일반적으로 모델의 최종 성능을 평가하는 데 사용되며, 모델을 개선하기 위해 추가적인 실험을 수행할 때 검증 데이터와 함께 사용됩니다. 테스트 데이터는 학습 및 검증 데이터와 동일한 분포를 가지며, 테스트 데이터를 미리 보고 모델을 조정하는 것은 일반화 오차를 과소평가할 가능성이 있습니다. 

따라서 테스트 데이터는 항상 마지막 단계에서만 사용되어야 하며, 모델 훈련 및 검증 단계에서는 절대 사용되어서는 안됩니다.

## *정형 데이터셋 나누기 실습*

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv'
df = pd.read_csv(url)
```

```python
columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'price' ]
df = df.loc[:,columns]

print(df)
```

```python
features_data = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors'] # 학습 데이터 
target_data = [ 'price' ] # 정답지 

x_data = df.loc[:,features_data] # 학습 데이터 
y_data = df.loc[:,target_data]   # 정답지 


print(x_data)
print(y_data)
```

### 데이터 나누기 train val test

```python
x_train, x_val_list, y_train, y_val_list = train_test_split(x_data, y_data, random_state=777, train_size=0.8)
x_val, x_test, y_val, y_test = train_test_split(x_val_list, y_val_list, random_state=777, test_size=0.5)
print("--------변경되기전 데이터 양--------------")
print("x_data 크기 : ", x_data.shape)
print("y_data 크기 : ", y_data.shape)
print("--------변경 후 데이터 양--------------")
print("x_train >> " , x_train.shape)
print("y_train >> ", y_train.shape)
print("x_val >> " , x_val.shape)
print("y_val >>" , y_val.shape)
print("x_test >>", x_test.shape)
print("y_test >> ", y_test.shape)
```
