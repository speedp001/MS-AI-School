```python
import numpy as np
from scipy.stats import wilcoxon

#그룹1 과 그룹2 데이터 생성
group1 = np.random.normal(0, 1, 100)
group2 = np.random.normal(0.5, 1, 100)

print(group2)
```

```python
#wilcoxon signed-rank test실행
st, p_val = wilcoxon(group1, group2)

#결과 출력
print("st : ", st)
print("p_val : ", p_val)
```
