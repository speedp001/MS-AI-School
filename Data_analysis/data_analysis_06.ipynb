```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#데이터 생성
np.random.seed(2)
x = np.random.normal(loc=0, scale=1, size=100)
y = 2 * x + np.random.normal(loc=0, scale=1, size=100)
print(y)
```

```python
# 선점도 시각화 
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
plt.show()
```
