```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. 이산형 데이터

이산형 데이터(Discrete data)는 정수 값으로 이루어진 데이터입니다. 이 데이터는 연속적이지 않으며, 한정된 개수의 값으로 이루어져 있습니다. 예를 들어, 동전 던지기 결과에서 나오는 앞면 또는 뒷면의 개수, 세대 수 등이 이산형 데이터의 대표적인 예시입니다.

### *막대 그래프*

```python
#데이터 읽어오기
data = pd.read_csv("./data/college_data.csv")
#print(data)
print(data['top10perc'].unique())   #unique()를 사용하면 Index값을 제외

# top10perc 막대 그래프 그리기
plt.figure(figsize=(10,6))
plt.bar(data['private'], data['top10perc'])
plt.title('Bar Chart of Top 10% Scores by Private/Public')
plt.xlabel('Private/Public')
plt.ylabel('Top 10% Scores')
plt.show()
```

### *원 그래프*

```python
# top10perc 원 그래프 그리기
plt.figure(figsize=(10,6))
data.groupby('private')['top10perc'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Top 10% Scores by Private/Public')
plt.ylabel('')
plt.show()
```

### *히스토그램*

```python
#top10perc 히스토그램 그리기
plt.hist(data['top10perc'], bins=20)
plt.title('Histogram of Top 10% Scores')
plt.xlabel('Top 10% Scores')
plt.ylabel('Frequency')
plt.show()
```

### *빈도수*

빈도수란 데이터에서 각 값이 나타나는 횟수를 의미합니다. 예를 들어, 주사위를 10번 던졌을 때, 1, 2, 3, 4, 5, 6 이 각 각 몇 번씩 나왔는지를 셀 수 있습니다. 이렇게 측정한 횟수가 빈도수입니다.

빈도수는 이산형 데이터를 분석할 때 중요한 개념 중 하나입니다. 이산형 데이터는 값이 유한하고 분리되어 있는 데이터이기 때문에, 각 값이 나타난 빈도수를 통해 전체 데이터의 분포를 파악할 수 있습니다.

```python
#데이터 생성
data = np.random.randint(1, 7, size=100)

#빈도수 계산
unique, counts = np.unique(data, return_counts=True)
#배열의 고유의 값과 그에 해당하는 반환 횟수를 반환

#막대 그래프 시각화
fig, ax = plt.subplots()
ax.bar(unique, counts)
ax.set_xlabel('Value')
ax.set_ylabel("Frequency")
ax.set_title("Frequency Distribution")
plt.show()
```

### *상대빈도수*

상대빈도수는 해당 데이터 값이 전체 데이터에서 차지하는 비율을 의미합니다. 즉, 각 데이터 값의 빈도를 전체 데이터 수로 나눈 값입니다. 예를 들어, 주사위를 10번 던져서 나온 결과가 1, 2, 3, 4, 5, 6 이 각각 1, 2, 1, 2, 2, 2 번 나왔다면, 각 값의 빈도수는 1, 2, 1, 2, 2, 2 이 됩니다. 전체 데이터 수가 10이므로, 각 값의 상대빈도수는 0.1, 0.2, 0.1, 0.2, 0.2, 0.2 가 됩니다.

상대빈도수는 빈도수와 달리, 전체 데이터의 크기에 상관 없이 각 데이터 값의 비율을 비교할 수 있으므로, 데이 터의 특성을 파악하는 데 유용합니다. 또한, 비율이기 때문에 항상 0과 1 사이의 값을 가지며, 이를 백분율로 변 환하여 해석하기도 쉽습니다.

```python
#예시 데이터 생성
data = np.array([1,2,2,2,2,2,3,4,5,6,7,7,7,8,9,10])
print(data)

#데이터에서 각 값의 빈도수 계산
value_counts = np.bincount(data)
print(value_counts)

print(len(data))
relative_frequencies = value_counts / len(data)
print(relative_frequencies)
```

```python
# 막대 그래프로 상대빈도수 시각화
plt.bar(np.arange(len(value_counts)), relative_frequencies)
plt.xticks(np.arange(len(value_counts)), np.arange(len(value_counts)))
plt.xlabel('Value')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Bar Chart')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D', 'F']
percentages = [30,40,15,10,5]

#원형그래프
fig, ax = plt.subplots()
ax.pie(percentages, labels=labels, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Percentages of Categories')
plt.show()
```
