```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# 학년별 학생 수
num_students_per_grade = 100

# 학년별로 학생를 분배
grade = np.repeat([1, 2, 3], num_students_per_grade)

# 각 학생의 과목별 점수
math_scores = np.random.uniform(0, 100, size=num_students_per_grade*3)
science_scores = np.random.uniform(0, 100, size=num_students_per_grade*3)
english_scores = np.random.uniform(0, 100, size=num_students_per_grade*3)

# 데이터화
data = {
    'Grade' : grade,
    'Math' : math_scores,
    'English' : english_scores,
    'Science' : science_scores
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 학년별 오름차순 정리
sorted_df = df.sort_values('Grade')
print(sorted_df)
```

```python
#폴더 생성
os.makedirs("./data", exist_ok=True)

#저장
df.to_csv("./data/student_scores.csv")
```

```python
#각 컬럼 상관계수 구하기
corr_matrix = df.corr()
print(corr_matrix)
```

```python
#히트맵 그리기
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
#각 점수간 상관관계가 거의 없다(상관계수가 거의 0에 수렴)
```

```python
#히스토그램을 그려주는 함수

def hist(df_name, df_columns) :
    plt.hist(df_name, bins=10, alpha=0.8)
    plt.xlabel('{} score'.format(df_columns))
    plt.ylabel('Frequency')
    plt.title('{} Histogram'.format(df_columns))
```

```python
hist(df.Math, df.columns[1])
```

```python
hist(df.English, df.columns[2])
```

```python
hist(df.Science, df.columns[3])
```
